$
\textbf{Previously:}\ \text{Independence of RVs}\\
\bullet \text{$X$ and $Y$ are independent if and only if}\\
\ \ \ \text{$F_{X,Y}(u,v) = F_X(u)F_Y(v)$ for all $u,v$}\\
\bullet \text{If $X$ and $Y$ are discrete-type, they are independent}\\
\ \ \ \text{if and only if $p_{X,Y}(u,v) = p_X(u)p_Y(v)$ for all $u,v$}\\
\bullet \text{If $X$ and $Y$ are jointly continuous-type, then they are independent}\\
\ \ \ \text{if and only if $f_{XY}(u,v) = f_X(u)f_Y(v)$ for all $u,v$}\\
\bullet \text{If $X$ and $Y$ are jointly continuous-type, then they are independent, if and only if,}\\
\ \ \ \text{for all $u$, either $f_X(u)=0$ or}\ 
\underbrace{f_{Y|X}(v|u) = f_Y(v)}_{\text{knowledge that $X=u$ doesn't change the distribution of $Y$}}
\ \text{for all $v$}
$

$
\textbf{Trick for deciding that two RVs are not independent}\\
\textbf{Definition}\ \text{$S\subset \mathbb{R}^2$ is a product set if it can be written as}\ S=A\times B = \{(u,v):\ u\in A,\ v\in B\}
$


$
\textbf{Note}:\ \text{$S$ is a product set if and only if it has the swap property}
$

$
\text{Proposition 1: If $X$ and $Y$ are jointly continuous-type, and their}\\
\text{support is not a product set, then $X$ and $Y$ are not independent.}\\
\text{Proposition 2: If $X$ and $Y$ are jointly uniformly distributed in $S$,}\\
\text{then $X$ and $Y$ are independent if and only if $S$ is a product set.}
$

$
\text{b) support is a product set}\\
\ \ \ \ \text{Let's compute $f_{Y|X}(v|u)$}\\
\ \ \ \ f_X(u) = \int_{0}^{1} (u+v)\ dv = uv + \frac{v^2}{2}|_0^1 = u + \frac{1}{2}\ (0\le u\le 1)\\
\ \ \ \ f_{Y|X}(v|u) = \frac{f_{XY}(u,v)}{f_X(u)} = \begin{cases}
\overbrace{\frac{u+v}{u+\frac{1}{2}}}^{\text{depends on $X=u$}} & 0\le v\le 1\\
0 & \text{otherwise}
\end{cases}\ (\text{defined for $0\le u\le 1$})\\
\ \ \ \ \Rightarrow \text{$X,Y$ are not independent}
$


$
\text{c) support is a product set}\\
\ \ \ \ f_X(u) = \begin{cases}
\int_0^1 9u^2v^2\ dv = 9u^2\cdot \frac{v^3}{3}|_0^1 = 3u^2 & 0 < u < 1\\
0 & \text{otherwise}
\end{cases}\\
\ \ \ \ f_Y(v) = \begin{cases}
3v^2 & 0 < v < 1\\
0 & \text{otherwise}
\end{cases}\\
\ \ \ \ f_X(u)f_Y(v) = \begin{cases}
9u^2v^2 & 0 < u < 1,\ 0 < v < 1\\
0 & \text{otherwise}
\end{cases} = f_{X,Y}(u,v)\\
\ \ \ \ \Rightarrow \text{$X$ and $Y$ are independent}
$

$
\textbf{Distribution of sums of RVs}\\
\ \\
\textbf{Sum of integer-valued RVs}\\
\text{Suppose $X$ and $Y$ are integer-valued discrete-type RVs.}\\
\text{Let}\ S = X+Y\\
p_S(k) = P\{X+Y=k\} = \sum_j P\{X=j, Y=k-j\} = \sum_j p_{XY}(j,k-j)\\
\text{If $X$ and $Y$ are independent,}\\
\sum_j p_{XY}(j,k-j) = \sum_j p_X(j)p_Y(k-j) = p_X * p_Y \rightarrow \text{convolution operation}
$

$
\text{$X$ is Binomial$(m,p)$, $Y$ is Binomial$(n,p)$, $X,Y$ are independent.}\\
p_S(k) = \sum_j p_X(j)p_Y(k-j) = \sum_j \binom{m}{j} p^j (1-p)^{m-j} \binom{n}{k-j} p^{k-j} (1-p)^{n-k+j}\\
= (\sum_j \binom{m}{j} \binom{n}{k-j}) p^k (1-p)^{m+n-k}
\rightarrow \text{first choose $j$ items from $m$ items, then choose $k-j$ items from $n$ items, for all values of $j$}\\
= \binom{m+n}{k} p^k (1-p)^{m+n-k} \rightarrow \text{# of subsets of size $k$ out of $m+n$ items, equivalently, pmf of Binomial$(m+n,p)$}
$

$
p_S(k) = \sum_j p_X(j)p_Y(k-j) = \sum_j \frac{e^{-\lambda_1} \lambda_1^j}{j!} \frac{e^{-\lambda_2} \lambda_2^{k-j}}{(k-j)!}\\
= \sum_j \frac{e^{-(\lambda_1+\lambda_2)}}{j! (k-j)!} \lambda_1^j \lambda_2^{k-j} \cdot \frac{k!}{k!}\\
= \frac{e^{-(\lambda_1+\lambda_2)}}{k!} \sum_j \binom{k}{j} \lambda_1^j \lambda_2^{k-j}\\
= \frac{e^{-(\lambda_1+\lambda_2)}}{k!} 
\underbrace{\sum_j \underbrace{\binom{k}{j}(\frac{\lambda_1}{\lambda_1+\lambda_2})^j (\frac{\lambda_2}{\lambda_1+\lambda_2})^{k-j}}_{\text{pmf of Binomial$(k,\frac{\lambda_1}{\lambda_1+\lambda_2})$}}}_{=1} \cdot (\lambda_1+\lambda_2)^k\\
= \frac{e^{-(\lambda_1+\lambda_2)} (\lambda_1+\lambda_2)^k}{k!} \rightarrow \text{Poisson$(\lambda_1+\lambda_2)$}
$
