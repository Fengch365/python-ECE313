\textbf{Ex 3.5.4}\ \text{Calls arrive at a Poisson Process with rate $\lambda$ = 2 calls/minute}\\
\text{($N_t$ = # of calls up to time t (measured in minutes) in Poisson$(\lambda t)$)}\\
\ \ \ \ \text{a)}\ P\{\text{no calls in first 3.5 minutes}\} = P\{\overbrace{N_{3.5}}^{\text{Poisson}(7)}=0\}
= \frac{e^{-7}\cdot 7^0}{0!} = e^{-7}\\
\ \ \ \ \ \ \ \ P\{\text{two or fewer calls in first 3.5 min}\}=P\{N_{3.5}\le 2\}
= P\{N_{3.5}=0\} + P\{N_{3.5}=1\} + P\{N_{3.5}=2\}\\
\ \ \ \ \ \ \ \ P\{\text{third call arrives after time t}\}=P\{N_t\le 2\} = P\{N_t= 0\} + P\{N_t= 1\} + P\{N_t= 2\}\\
\ \ \ \ \ \ \ \ = e^{-2t} + \frac{e^{-2t} (2t)}{1!} + \frac{e^{-2t} (2t)^2}{2!} = e^{-2t} (1 + 2t + 2t^2)\\
\ \ \ \ \text{b) Find pdf of time of third call ($T_3$)}\\
\ \ \ \ \ \ \ \ \text{First, find CDF:}\ F_{T_3}(t) = P\{T_3\le t\} = 1-P\{T_3>t\}=1-e^{-2t}(1 + 2t + 2t^2)\\
\ \ \ \ \ \ \ \ \text{Find}\ \frac{d}{dt} F_{T_3}(t) = 2e^{-2t} (1 + 2t + 2t^2) - e^{-2t}(2+4t) = e^{-2t} (2t)^2\\
\ \ \ \ \text{c) Find expected time of tenth arrival}\ E[T_{10}]\\
\ \ \ \ \ \ \ \ T_{10} = U_1 + U_2 +\ ...\ + U_{10} \Rightarrow E[T_{10}] = 10 E[U_i] = \frac{10}{\lambda} = 5\ \text{min}\\
$

\textbf{Erlang distribution}\\
\ \\
\text{Let $T_r$ be the time of $r^{\text{th}}$ count in a Poisson Process with rate $\lambda$.}\\
\text{CDF of $T_r$:}\ P\{T_r\le t\} = 1-P\{T_r>t\} = 1- P\{N_t\le r-1\}= 1- \sum_{k=0}^{r-1} P\{N_t=k\}
= 1- \sum_{k=0}^{r-1} \frac{e^{-\lambda t} (\lambda t)^k}{k!} \\
\text{pdf obtained through differentiation:}\ f_{T_r}(t) = \frac{d}{dt} (1- \sum_{k=0}^{r-1} \frac{e^{-\lambda t} (\lambda t)^k}{k!})
= \underbrace{\frac{e^{-\lambda t} \lambda^r (t)^{r-1}}{(r-1)!}}_{\text{Erlang's distribution}}\\
$

$
\textbf{Linear scaling of pdfs}\\
\ \\
\text{Suppose $X$ has pdf $f_X(x)$, and $Y=aX+b$, for $a>0$. What is pdf of $Y$?}\\
\ \ \ \ \text{First, find CDF:}\ F_Y(y) = P\{Y\le y\} = P\{aX+b\le y\} = F_X(\frac{y-b}{a})\\
\ \ \ \ \Rightarrow f_Y(y) = \frac{d}{dy} F_Y(y) = \frac{d}{dy}F_X(\frac{y-b}{a}) = f_X(\frac{y-b}{a})\cdot \frac{d}{dy}(\frac{y-b}{a})\\
\ \ \ \ \Rightarrow f_Y(y) = \frac{1}{a} f_X(\frac{y-b}{a}),\ \text{for $a>0$}\\
\ \ \ \ \Rightarrow f_Y(y) = -\frac{1}{a} f_X(\frac{y-b}{a}),\ \text{for $a<0$}\\
$

\textbf{Ex 3.6.2}\ \text{The waiting time $T$ for a service has Exponential$(0.01)$ distribution,}\\
\text{when measured in seconds. Let $S$ be the same time, but in minutes.}\\
\ \ \ \ \text{pdf of $T$:}\ f_T(t) = \begin{cases}
0.01 e^{-0.01 t} & t\ge 0\\
0 & t<0
\end{cases}\\
\ \ \ \ S=\frac{T}{60}\ (a=\frac{1}{60},\ b=0)\\
\ \ \ \ \text{pdf of $S$}\ f_S(s) = 60 f_T(60s) = \begin{cases}
0.6 e^{-0.6s} & s\ge 0\\
0 & s<0
\end{cases}
