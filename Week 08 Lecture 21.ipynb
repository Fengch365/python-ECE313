$
\textbf{Poisson Process}\\
\ \\
\text{Recall: Bernoulli process $X_1, X_2, X_3, ...$, all independent Bernoulli$(p)$ RVs}\\
\ \ \ \ \text{$L_i$'s are inter-count times. $L_i$ is Geometric$(p)$}\\
\ \ \ \ \text{Count at time k:}\ C_k = X_1+X_2+\ ...\ +X_k\Rightarrow \text{$C_k$ is Binomial$(k,p)$}
$


$
\text{Intuition for Poisson Process.: Let time between Bernoulli trials be $h$, and consider $h$ very small, $h\to 0$ (and $p\to 0$, with $\frac{p}{h}=\lambda$)}\\
\text{Poisson Process: events can happen at any real-valued t}\\
\ \ \ \ \text{$U_i$'s are inter-count times}\\
\ \ \ \ \text{$U_i$'s are Exponential$(\lambda)$ and independent}\\
\ \ \ \ \text{Count at time $t$: $N_t$ is Poisson$(\lambda t)$}
$

$
\textbf{Formal definition}\\
\text{Let $\overbrace{\lambda}^{\text{rate}} > 0$. A Poisson Process with rate $\lambda$ is a counting process $N=(N_t \ t\ge 0)$ such that}\\
\ \ \ \ \text{N1) For $0\le t_1\le t_2\le t_3\le \ ...\ \le t_n$, the increments $N_{t_1}-N_{0},\ N_{t_2}-N_{t_1},\ N_{t_3}-N_{t_2},\ 
...,\ N_{t_n}-N_{t_{n-1}}$ are independent}\\
\ \ \ \ \text{N2) Increment $N_t-N_s$ has Poisson$(\lambda(t-s))$ distribution}\\
\ \\
\text{Poisson Process often models customers arriving in a store or packets arriving in a router, etc.}
$

$
\textbf{Ex 3.5.3}\ \text{Poisson Process with rate $\lambda >0$ in $[0,T]$. Fix $\tau \in [0.T]$}\\
\text{Let $X_1$ be counts in $[0,\tau]$, $x_2$ be counts in $(\tau,T]$, $\overbrace{X}^{\text{total counts}} = X_1+X_2$}\\
\ \ \ \ \text{a)}\ P\{X=n\} = \frac{e^{-\lambda T} (\lambda T)^n}{n!}\ \text{($X$ is Poisson$(\lambda T)$)}\\
\ \ \ \ \text{b)}\ P\{X_1 = i\} = \frac{e^{-\lambda \tau} (\lambda \tau)^i}{i!}\ \text{($X_1$ is Poisson$(\lambda \tau)$)}\\
\ \ \ \ \text{c)}\ P\{X_2=j\} = \frac{e^{-\lambda (T-\tau)} (\lambda (T-\tau))^j}{j!}\ \text{($X_2$ is Poisson$(\lambda (T-\tau))$)}\\
\ \ \ \ \text{d)}\ \text{For $i<n$,}\ P(X_1=i|X=n) = \frac{P(X_1=i,X=n)}{P(X=n)}=\frac{P(X_1=i,X_2=n-i)}{P(X=n)}\ \text{($X_1,X_2$ are independent)}\\
\ \ \ \ \ \ \ \ = \frac{P(X_1=i)P(X_2=n-i)}{P(X=n)} = \frac{e^{-\lambda \tau} (\lambda \tau)^i}{i!} \cdot \frac{e^{-\lambda(T-\tau)} (\lambda(T-\tau))^{n-i}}{(n-i)!} \cdot \frac{n!}{e^{-\lambda T} (\lambda T)^n}\\
\ \ \ \ \ \ \ \ = \frac{n!}{i!(n-i)!} \underbrace{e^{-\lambda(\tau+T-\tau -T)}}_{=1}\frac{\lambda^n \tau^i (T-\tau)^{n-i}}{\lambda^n T^i T^{n-i}} = \underbrace{\binom{n}{i} (\frac{\tau}{T})^i (1-\frac{\tau}{T})^{n-i}}_{\text{pmf of Binomial$(n,\frac{\tau}{T})$}}\\
\ \ \ \ \text{Intuition: Conditioned on $X=n$, each count falls in $(0,\tau)$ with probability $\frac{\tau}{T}$}\\
\ \ \ \ \text{e)}\ P(X=n|X_1=i) = P(X_2=n-i|X_1=i) = P(X_2=n-i) = \frac{e^{-\lambda (T-\tau)} (\lambda (T-\tau))^{n-i}}{(n-i)!}
$
