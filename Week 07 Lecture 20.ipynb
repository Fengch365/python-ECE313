$
\textbf{Previously}\ \text{Continuous-type RVs}\\
\bullet \text{pdf $f_X(x)$, CDF $F_X(c) = \int_{-\infty}^{c} f_X(u)\ du$}\\
\bullet f_X(x) = \frac{d}{dx} F_X(x)\ \text{(where $F_X$ is differentiable)}\\
$

$
\bullet\ \text{Expectation}\ E[X] = \int_{-\infty}^{\infty} x f_X(x)\ dx\ \text{(Reccall for discrete-type $E[X] = \sum_{u_i} u_i p_X(u_i)$)}\\
\ \ \ \ \bullet \text{Linearity still holds, e.g.}\ E[\alpha X+\beta]=\alpha E[X] + \beta\\
\bullet \text{LOTUS}\ E[g(X)] = \int_{-\infty}^{\infty} g(x) f_X(x)\ dx\\
\bullet \text{Var}(X) = E[(X-\mu_X)^2] = E[X^2] - E[X]^2\\
$

$
\textbf{Uniform Distribution}\\
\ \\
\text{A RV $X$ is uniformly distributed in $[a,b]$ if its pdf is}\
f_X(u) = \begin{cases}
\frac{1}{b-a} & \text{if}\ a\le u\le b\\
0 & \text{otherwise}
\end{cases}\\
\bullet E[X] = \frac{a+b}{2}\\
\bullet E[X^2] = \int_{-\infty}^{\infty} u^2 f_X(u)\ du = \int_a^b \frac{u^2}{b-a}\ du = \frac{1}{b-a} \frac{u^3}{3}|_a^b 
= \frac{b^2+ab+a^2}{3}\\
\Rightarrow \text{Var}(X) = E[X^2]-E[X]^2 = \frac{(b-a)^2}{12}
$


$
\textbf{Exponential Distribution}\\
\ \\
\text{A RV $T$ has the exponential distribution with parameter $\lambda>0$, if its pdf is}\ 
f_T(t) = \begin{cases}
\lambda e^{-\lambda t} & t\ge 0\\
0 & \text{otherwise}
\end{cases}
$


$
\bullet\ \text{CDF}\ F_T(t) = \int_{-\infty}^t f_T(u)\ du = \int_0^t \lambda e^{-\lambda u}\ du = -e^{-\lambda u}|_0^t = 1-e^{-\lambda t}\\
\bullet E[T^n] = \int_{-\infty}^{\infty} t^n f_T(t)\ dt 
= \int_0^{\infty} \underbrace{t^n}_{u} \underbrace{\lambda e^{-\lambda t}\ dt}_{dv} 
\overbrace{=}^{\text{integration by parts}} \underbrace{t^n\cdot (-1)\cdot e^{-\lambda t}|_0^{\infty}}_{0} + \int_0^{\infty} (nt^{n-1}) e^{-\lambda t}\ dt\\
\ \ \ \ \ \ \ \ \ \ \ \ \
= \frac{n}{\lambda} \int_0^{\infty} t^{n-1} \overbrace{(\lambda e^{-\lambda t})}^{\text{pdf}}\ dt
= \frac{n}{\lambda} E[T^{n-1}] \underbrace{=}_{\text{recurse}} \frac{n}{\lambda} \frac{n-1}{\lambda} E[T^{n-2}]
=\ ...\ = \frac{n!}{\lambda^n} \overbrace{E[T^0]}^{=1} = \frac{n!}{\lambda^n}\\
\text{In particular,}\ E[T]=\frac{1}{\lambda},\ E[T^2]=\frac{2}{\lambda^2}\\
\Rightarrow \text{Var}(T) = \frac{2}{\lambda^2} - (\frac{1}{\lambda})^2 = \frac{1}{\lambda^2}\\
\bullet \text{Memoryless Property of Exponential distribution}\\
\ \ \ \text{If T is exponential with parameter $\lambda$,}\ P\{T>t\} = 1-F_T(t) = e^{-\lambda t}\ 
\text{(Note: $F_T(t)=1-e^{-\lambda t}$)}\\
\ \ \ \text{Hence, for}\ s,t > 0,\ P(T>s+t|T>s) = \frac{P(T>s+t,T>s)}{P(T>s)} = \frac{e^{-\lambda(t+s)}}{e^{-\lambda s}}
=e^{-\lambda t} = P\{T>t\}\\
\ \\
