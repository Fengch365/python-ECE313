$
F_R(c) = P(R\le c) = P(X^2 + Y^2\le c^2)\\
= \iint_D f_{XY} (u,v)\ dvdu = \int_{\theta=0}^{2\pi} \int_{r=0}^{c}\ f_{XY} (r\ cos(\theta), r\ sin(\theta))r\ drd\theta
$


$
f_R(c) = \frac{d}{dc} F_R(c) = \int_0^{2\pi} \frac{d}{dc} (\int_{r=0}^{c}\ f_{XY} (r\ cos(\theta), r\ sin(\theta)) r\ dr)\ d\theta\\
= \int_{0}^{2\pi} f_{XY}(c\ cos(\theta), c\ sin(\theta))c\ d\theta\\
\text{special case:}\ f_{XY}(u,v)\ \text{only depends on}\ u^2+v^2\ \text{(e.g. $f_{XY}(u,v) = \frac{exp(-\frac{u^2+v^2}{2})}{2\pi}$)}\\
\Rightarrow f_{XY}(\underbrace{c\ cos(\theta)}_{x}, \underbrace{c\ sin(\theta)}_{y}) = f_{XY}(c\cdot cos(0), c\cdot sin(0)) = f_{XY}(c,0)\\
\because (c\ cos(\theta))^2 + (c\ sin(\theta))^2 = c^2\\
\therefore f_R(c) = \int_{0}^{2\pi} f_{XY}(c,0)c\ d\theta = f_{XY}(c,0)c\cdot \theta |_{0}^{2\pi} = (2\pi \cdot c) f_{XY}(c,0)
$

$
\text{needle intersects line if}\ sin(\Theta) > U\\
P(\text{needle intersets line}) = P(U< sin(\Theta)) = \int_{0}^{\pi} \int_{0}^{sin (\theta)} \frac{1}{\pi}\ dud\theta\\
= \int_{0}^{\pi} \frac{u}{\pi}|_{0}^{sin (\theta)}\ d\theta = \int_{0}^{\pi} \frac{sin (\theta)}{\pi}\ d\theta
= -\frac{cos(\theta)}{\pi} |_{0}^{\pi}\\
= \frac{1-(-1)}{\pi} = \frac{2}{\pi} = 0.637
$

$
\text{Jointly uniformly distributed}\ [0,\pi]\times [0,1]\times [0,1]\\
P(\text{needle doesn't intersect any line}) = \\
P(\text{needle doesn't intersect any horizontal line, doesn't intersect any vertical line}) = \\
P(\underbrace{U > sin(\Theta), V > |cos(\Theta)|}_{\text{not independent}\to \text{because}\ cos(\theta)<0\ \text{for}\ \theta \in 
(\frac{\pi}{2}, \pi)})\\
= \int_{\theta=0}^{\pi} \int_{u=sin(\theta)}^{1} \int_{v=|cos(\theta)|}^{1} \frac{1}{\pi}\ dvdud\theta
= \int_{\theta=0}^{\pi} \frac{v}{\pi}|_{|cos(\theta)|}^{1} u|_{sin(\theta)}^{1}\ d\theta\\
= \int_{\theta=0}^{\pi} \frac{(1-sin(\theta))(1-|cos(\theta)|)}{\pi}\ d\theta\\
= \int_{0}^{\frac{\pi}{2}} \frac{1-cos(\theta)-sin(\theta)+sin(\theta)cos(\theta)}{\pi}\ d\theta + 
\int_{\frac{\pi}{2}}^{\pi} \frac{1+cos(\theta)-sin(\theta)-\overbrace{sin(\theta)cos(\theta)}^{\frac{1}{2}sin(2\theta)}}{\pi}\ d\theta\\
= 1 - \frac{3}{\pi} \approx 0.045
$
