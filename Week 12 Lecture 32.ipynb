$
\textbf{Previously}\\
\text{Distribution of sums of RVs}\\
\bullet \text{sum of integer-valued discrete-type RVs $X$, $Y$ are integer-valued RVs. $S=X+Y$}\\
\ \ \ p_S(k) = \sum_j p_{XY} (j,k-j)\\
\ \ \ \text{If $X$ and $Y$ are independent,}\\
\ \ \ p_S(k) = \sum_j p_X(j) p_Y(k-j) = p_X \underbrace{\ast}_{\text{discrete-time convolution}} p_Y
$

$
\textbf{Sums of continuous-type RVs}\\
\text{Suppose $X$, $Y$ are jointly continuous-type (joint pdf $f_{XY}$)}\\
\text{Let $S=X+Y$. What is $f_S(c)$?}\\
$


$
F_S(c) = P(S\le c) = P(X+Y\le c)= P(Y\le c-X) = \int_{-\infty}^{\infty} \int_{-\infty}^{c-u} f_{XY}(u,v)\ dv du\\
f_S(c) = \frac{d}{dc} F_S(c) = \frac{d}{dc} \int_{-\infty}^{\infty}(\int_{-\infty}^{c-u} f_{XY}(u,v)\ dv)\ du
= \int_{-\infty}^{\infty} \frac{d}{dc} (\int_{-\infty}^{c-u} f_{XY}(u,v)\ dv)\ du\\
= \int_{-\infty}^{\infty} \frac{d}{dc} (B(c-u)-B(-\infty))\ du = \int_{-\infty}^{\infty} f_{XY}(u,c-u)\ du\\
\Rightarrow f_S(c) = \int_{-\infty}^{\infty} f_{XY}(u,c-u)\ du\ \text{(compare with $p_S(k)=\sum_j p_{XY}(j,k-j)$)}\\
\text{If $X$ and $Y$ are independent,}\ f_S(c) = \int_{-\infty}^{\infty} f_X(u)f_Y(c-u)\ du 
= f_X \underbrace{\ast}_{\text{continuous-time convolution}} f_Y
$

$
f_X(u) = \begin{cases}
1 & 0<u<1\\
0 & \text{otherwise}
\end{cases},\ \ 
f_Y(v) = \begin{cases}
1 & 0<v<1\\
0 & \text{otherwise}
\end{cases}\\
f_S(c) = \int_{-\infty}^{\infty} f_X(u)f_Y(c-u)\ du\
\text{, where}\
f_Y(c-u) = \begin{cases}
1 & c-1<u<c\\
0 & \text{otherwise}
\end{cases}
$


$
\text{For}\ 1<c<2:\ f_S(c) = \int_{-\infty}^{\infty} f_X(u)f_Y(c-u)\ du = \int_{c-1}^{1} 1\ du = u|_{c-1}^{1} = 2-c\\
\text{For}\ 0<c<1:\ f_S(c) = \int_{0}^{c} 1\ du = u|_{0}^{c} = c\\
\text{otherwise,}\ f_S(c) = 0
$


$
f_S(c) = \int_{-\infty}^{\infty} f_X(u) f_Y(c-u)\ du 
= \int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi \sigma_1^2}} exp(-\frac{u^2}{2\sigma_1^2})
\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi \sigma_2^2}} exp(-\frac{(c-u)^2}{2\sigma_2^2})\ du\\
= \frac{1}{2\pi \sqrt{\sigma_1^2 \sigma_2^2}} \int_{-\infty}^{\infty} 
exp(-\frac{u^2}{2\sigma_1^2}-\frac{(c-u)^2}{2\sigma_2^2})\ du\\
\text{where}\ -\frac{u^2}{2\sigma_1^2}-\frac{(c-u)^2}{2\sigma_2^2}
= -\frac{1}{2} (\frac{u^2}{\sigma_1^2}+\frac{c^2-2cu+u^2}{\sigma_2^2})
= -\frac{1}{2} (\underbrace{(\frac{1}{\sigma_1^2}+\frac{1}{\sigma_2^2})}_{A}\cdot u^2 
- \underbrace{\frac{2c}{\sigma_2^2}}_{B}\cdot u
+ \underbrace{\frac{c^2}{\sigma_2^2}}_{C})\\
\because Au^2-Bu+C = A(u^2 - \frac{B}{A} u + \frac{C}{A}) = A(u^2-\frac{B}{A}u + \frac{B^2}{4A^2} - \frac{B^2}{4A^2} + \frac{C}{A})
= A(u-\frac{B}{2A})^2 - \frac{B^2}{4A^2} + C\\
\therefore f_S(c) = 
\frac{1}{2\pi \sqrt{\sigma_1^2 \sigma_2^2}} \int_{-\infty}^{\infty} exp(-\frac{A}{2}(u-\frac{B}{2A})^2)
\cdot \underbrace{exp(-\frac{1}{2}(C-\frac{B^2}{4A}))}_{\text{doesn't depend on $u$}}\ du\\
\ \ \ \ = \frac{1}{2\pi \sqrt{\sigma_1^2 \sigma_2^2}} exp(-\frac{1}{2}(C-\frac{B^2}{4A}))\cdot
\sqrt{2\pi \cdot \frac{1}{A}} 
\underbrace{\int_{-\infty}^{\infty} 
\underbrace{\frac{exp(-\frac{(u-\frac{B}{2A})^2}{2(\frac{1}{A})})}{\sqrt{2\pi \cdot \frac{1}{A}}}}_{\text{pdf of $N(\frac{B}{2A}, \frac{1}{A})$}}\ du}_{=1}
= \underbrace{\frac{1}{\sqrt{2\pi (\sigma_1^2 + \sigma_2^2)}} exp(-\frac{c^2}{2 (\sigma_1^2 + \sigma_2^2)})}
_{\text{pdf of $N(0,\ \sigma_1^2 + \sigma_2^2)$}}
$

\textbf{Additional problem on joint distribution}

$
\text{Find CDF first}\\
F_W(c) = P(W\le c) = P(max(X,Y)\le c)= P(X\le c,\ Y\le c)\\
= P(X\le c)P(Y\le c) = F_X(c) F_Y(c)\\
f_W(c) = \frac{d}{dc} F_W(c) = (\frac{d}{dc} F_X(c))F_Y(c) + F_X(c)(\frac{d}{dc} F_Y(c))\\
= f_X(c)F_Y(c) + F_X(c)f_Y(c)
