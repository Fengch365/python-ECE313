\textbf{The Gaussian (Normal) distribution}\\
\ \\
\text{Notation:}\ N(\mu, \sigma^2)\\
\text{pdf:}\ f(u) = \frac{1}{\sqrt{2\pi \sigma^2}} exp(-\frac{(u-\mu)^2}{2\sigma^2})\\
\text{mean:}\ E[X]=\mu\\
\text{variance:}\ \text{Var}(X) = \sigma^2\\
\ \\
$


$
\textbf{Standard Normal Distribution}\\
\ \\
N(0,1)\ \text{distribution}\\
\text{CDF of $N(0,1)$:}\ \Phi (u) = P\{X\le u\} = \int_{-\infty}^{u} \frac{1}{\sqrt{2\pi}} e^{-\frac{v^2}{2}}\ dv\\
\text{Complementary CDF of $N(0,1)$:}\ Q(u)=P\{X>u\}=\int_{u}^{\infty} \frac{1}{\sqrt{2\pi}} e^{-\frac{v^2}{2}}\ dv\\
\textbf{Note}\ Q(u) = 1-\Phi(u)=\Phi(-u)\\
\textbf{Important}\ \text{If $Y$ has $N(\mu, \sigma^2)$ distribution, then $Y=\sigma X + \mu$, where $X$ is $N(0,1)$}
$

$
\textbf{Ex 3.6.4}\ \text{Let $X$ have $N(10,16)$ distribution. Compute:}\\
\ \ \ \ \text{a)}\ P\{X\ge 15\}\\
\ \ \ \ \ \ \ \ \because X = 4\cdot Z + 10,\ \text{where $Z$ is $N(0,1)$} \Rightarrow Z= \underbrace{\frac{X-10}{4}}_{N(0,1)}\\
\ \ \ \ \ \ \ \ \therefore P\{X\ge 15\} = P\{X-10\ge 5\}=P\{\frac{X-10}{4}\ge \frac{5}{4}\} = Q(\frac{5}{4})\\
\ \ \ \ \text{b)}\ P\{X\le 5\} = P\{\underbrace{\frac{X-10}{4}}_{N(0,1)}\le -\frac{5}{4}\} = \Phi(-\frac{5}{4}) = Q(\frac{5}{4})\\
\ \ \ \ \text{c)}\ P\{X^2\ge 400\} = P\{X\ge 20\ \text{or}\ X\le -20\} = P\{X\ge 20\} + P\{X\le -20\}
= P\{\frac{X-10}{4}\ge \frac{10}{4}\} + P\{\frac{X-10}{4}\le -\frac{30}{4}\}\\
\ \ \ \ \ \ \ \ = Q(2.5)+\Phi(-7.5) = Q(2.5) + Q(7.5) \approx Q(2.5)\\
\ \ \ \ \text{d)}\ P\{X=2\} = 0\ \text{($X$ is continuous-type)}
$

\textbf{Central Limit Theorem}\\
\ \\
\textbf{Main idea} \text{: sum of many independent RVs, all of similar variances, is approximately Gaussian}\\
\text{We consider first the case of Bernoulli$(p)$ RVs, $X_1,X_2,X_3,...$}\\
\text{Let $S_{n,p}=\sum_{i=1}^{n} X_i$.}\\
\text{$S_{n,p}$ is Binomial$(n,p)$, so $E[S_{n,p}]=np$ and Var$(S_{n,p})=np(1-p)$}\\
\text{"standardized" version: $\frac{S_{n,p}-np}{\sqrt{np(1-p)}}$}\\
\ \\
\textbf{DeMoivre-Laplace Limit Theorem}\\
\text{Suppose $S_{n,p}$ is a binomial random variable with parameters $n$ and $p$}\\
\text{For any fixed $c$,}\ lim_{n\to \infty} P\{\frac{S_{n,p}-np}{\sqrt{np(1-p)}} \le c\} = \Phi(c)\\
\text{(Intuitively, $\frac{S_{n,p}-np}{\sqrt{np(1-p)}}$ is approximately a $N(0,1)$ RV)}
