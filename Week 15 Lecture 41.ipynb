$
\textbf{Previously:}\ \text{Joint Gaussian Distribution (bivariate normal)}\\
\textbf{Definition:}\ \text{RVs $X$ and $Y$ are jointly Gaussian if $aX+bY$ is Gaussian for all $a,b\in \mathbb{R}$}\\
\text{As long as $\rho = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y} \neq 1$ and $\rho \neq -1$, $X$ and $Y$ have a joint pdf,}\\
\text{given by}\ f_{XY}(u,v) = \frac{1}{2\pi \sigma_X \sigma_Y \sqrt{1-\rho^2}} 
exp(-\frac{(\frac{u-\mu_X}{\sigma_X})^2 + (\frac{v-\mu_Y}{\sigma_Y})^2 - 2\rho (\frac{u-\mu_X}{\sigma_X})(\frac{v-\mu_Y}{\sigma_Y})}{2(1-\rho^2)}),\\
\mu_X = E[X],\ \mu_Y = E[Y], {\sigma_X}^2 = \mbox{Var}(X),\ {\sigma_Y}^2 = \mbox{Var}(Y),\ \rho = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}
$

$
\textbf{Key properties of bivariate normal}\\
\text{Proposition:}\ \text{Suppose $X,Y$ have bivariate normal distribution with}\ \rho = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}\\
\text{d) $X$ and $Y$ are independent if and only if $\rho=0$, i.e.,}\\
\ \ \ \ \text{for jointly Gaussian RVs, uncorrelated $\iff$ independent}\\
\text{e) The MMSE estimator of $Y$ given $X$ is linear.}\\
\ \ \ \ E[Y|X] = \hat{E}[Y|X] = E[Y] + \frac{\text{Cov}(X,Y)}{\text{Var}(X)} (X-E[X])\\
\text{f) The conditional distribution of $Y$ given $X=u$ is}\ N(\hat{E}[Y|X=u], {\sigma_e}^2),\\
\ \ \ \ \text{where}\ {\sigma_e}^2 = E[(Y-\hat{E}[Y|X])^2] = \text{Var}(Y) - \frac{\text{Cov}(X,Y)^2}{\text{Var}(X)}\\
\ \ \ \ \text{In particular, if $X,Y$ are $N(0,1)$ and their correlation coefficient is $\rho$, then}\\
\ \ \ \ \text{given $X=u$, the conditional distribution of $Y$ is $N(\rho u, 1-\rho^2)$}
$

$
\textbf{Higher dimension joint Gaussian distribution}\\
\text{For a random vector $X=\begin{bmatrix} X_1\\ X_2\\ \vdots \\ X_n \end{bmatrix}$, the n-dimensional Gaussian density is given by}\\
f_{X_1X_2...X_n}(\underbrace{\vec{u}}_{\vec{u}=\begin{bmatrix} u_1\\ \vdots \\ u_n \end{bmatrix}}) 
= \frac{1}{(2\pi)^{\frac{n}{2}} \sqrt{\text{det}\Sigma}} 
exp(-\frac{(\vec{u}-\vec{\mu})^T\ \Sigma^{-1}\ (\vec{u}-\vec{\mu})}{2}),\ \mbox{where}\ \vec{\mu} = \begin{bmatrix} E[X_1]\\ E[X_2]\\ \vdots \\ E[X_n]
\end{bmatrix}\ \text{is the vector of means and}\\
\Sigma\ \text{is the $n\times n$ covariance matrix}\ \Sigma = \begin{bmatrix}
\text{Cov}(X_1,X_1) & \text{Cov}(X_1,X_2) & ... & \text{Cov}(X_1,X_n)\\
\text{Cov}(X_2,X_1) & \text{Cov}(X_2,X_2) & ... & \vdots \\
... & ... & \ddots & \vdots \\
\text{Cov}(X_n,X_1) & \text{Cov}(X_n,X_2) & ... & \text{Cov}(X_n,X_n)
\end{bmatrix}\\
\text{special cases:}\\
\bullet \mu = \begin{bmatrix} 0\\ 0\\ \vdots \\ 0 \end{bmatrix}\ \mbox{and}\ \Sigma = \begin{bmatrix}
1 & 0 & ... & 0\\
0 & 1 & ... & \vdots \\
... & ... & \ddots & \vdots \\
0 & ... & ... & 1
\end{bmatrix} = I,\ \mbox{det}\Sigma = 1,\ \Sigma^{-1} = I,\ \mbox{and}\\
\ \ \ f_{X_1...X_n}(\vec{u}) = \frac{1}{(2\pi)^{\frac{n}{2}} \cdot 1} exp(-\frac{\vec{u}^T\ I\ \vec{u}}{2})
= \frac{1}{(2\pi)^{\frac{n}{2}}} exp(-\frac{{u_1}^2 + {u_2}^2 + ... + {u_n}^2}{2})\\
\ \ \ = \prod_{i=1}^{n} (\frac{1}{\sqrt{2\pi}} e^{-\frac{{u_i}^2}{2}}) \implies X_1,X_2,...,X_n\ \text{are independent $N(0,1)$ RVs}\\
\bullet n=1,\ \Sigma = \sigma^2,\ \mbox{det} \Sigma = \sigma^2,\ \Sigma^{-1} = \frac{1}{\sigma^2}\ \mbox{and}\
f_X(u) = \frac{1}{\sqrt{2\pi \sigma^2}} exp(-\frac{(u-\mu)^2}{2\sigma^2})\\
\bullet n=2,\ \Sigma = \begin{bmatrix}
\text{Cov}(X,X) & \text{Cov}(X,Y)\\
\text{Cov}(Y,X) & \text{Cov}(Y,Y)
\end{bmatrix} = \begin{bmatrix}
{\sigma_X}^2 & \rho \sigma_X \sigma_Y\\
\rho \sigma_X \sigma_Y & {\sigma_Y}^2
\end{bmatrix}\\
\ \ \ \mbox{det} \Sigma = {\sigma_X}^2 {\sigma_Y}^2 - \rho^2 {\sigma_X}^2 {\sigma_Y}^2 = {\sigma_X}^2 {\sigma_Y}^2 (1-\rho^2)\\
\ \ \ \Sigma^{-1} = \frac{1}{{\sigma_X}^2 {\sigma_Y}^2 (1-\rho^2)} \begin{bmatrix}
{\sigma_Y}^2 & -\rho \sigma_X \sigma_Y\\
-\rho \sigma_X \sigma_Y & {\sigma_X}^2
\end{bmatrix}\\
\ \ \ \text{In the exponent of the joint pdf we get}\\
\ \ \ (\vec{u}-\vec{\mu})^T \Sigma^{-1} (\vec{u}-\vec{\mu}) = -\frac{1}{1-\rho^2}
((\frac{u_1-\mu_X}{\sigma_X})^2 + (\frac{u_2-\mu_Y}{\sigma_Y})^2 - 2\rho (\frac{u_1-\mu_X}{\sigma_1})(\frac{u_2-\mu_Y}{\sigma_2}))
$
