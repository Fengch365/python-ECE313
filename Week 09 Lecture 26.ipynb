\bullet \text{Failure Rate Functions $h(t)$}\\
\ \ \ \text{Describe the "rate" at which an object fails as a function of time (e.g. a light bulb)}
$

$
\ \ \ \text{Let $T$ be the time until failure.}\\
\ \ \ \text{Failure rate function can be computed as}\ h(t) = \frac{f_T(t)}{1-F_T(t)}\\
\ \ \ \text{Conversely, if we are given $h(t)$, can find $F_T(t)$.}\\
\ \ \ \text{To do that, we notice that}\ h(t) = \frac{f_T(t)}{1-F_T(t)} = \frac{F_T'(t)}{1-F_T(t)} = -(\ln (1-F_T(t)))'\\
\ \ \ -\int_0^t h(s)\ ds = + \ln (1-F_T(t))\\
\ \ \ e^{-\int_0^t h(s)\ ds} = 1-F_T(t) \Rightarrow F_T(t) = 1 - e^{-\int_0^t h(s)\ ds}
$

$
\text{a) $T$ is Exponential$(\lambda)$.}\ f_T(t) = \lambda e^{-\lambda t}\ \text{for}\ t\ge 0\\
\ \ \ \ F_T(t) = 1 - e^{-\lambda t},\ h(t) = \frac{f_T(t)}{1-F_T(t)} = \frac{\lambda e^{-\lambda t}}{1-(1-e^{-\lambda t})} = \lambda\\
\ \ \ \ \text{exponential time to failure $\Leftrightarrow$ constant failure rate}\\
\ \ \ \ \text{(think of the memoryless property)}
$

$
\text{b)}\ h(t) = \frac{t}{\sigma^2}\ \text{for}\ t\ge 0\\
\ \ \ \ F_T(t) = 1 - e^{-\int_{0}^{t} h(s)\ ds},\ \int_{0}^{t} h(s)\ ds = \int_{0}^{t} \frac{s}{\sigma^2}\ ds
= \frac{s^2}{2\sigma^2} |_{0}^{t} = \frac{t^2}{2\sigma^2}\\
\ \ \ \ \Rightarrow F_T(t) = 1 - e^{-\frac{t^2}{2\sigma^2}},\ f_T(t) = \frac{d}{dt} F_T(t) = \frac{t}{\sigma^2} e^{-\frac{t^2}{2\sigma^2}}\
\text{for}\ t\ge 0\\
\text{c)}\ \overbrace{T}^{\text{computer fails}} = \text{min}(\overbrace{T_1}^{\text{motherboard}},\overbrace{T_2}_{\text{screen}})\\
\ \ \ \ P(T > t) = P(\text{min}(T_1,T_2) > t) = P(T_1>t, T_2>t) \underbrace{=}_{\text{independent}} P(T_1>t)P(T_2>t)\\
\ \ \ \ \Rightarrow 1-F_T(t) = (1-F_{T_1}(t))(1-F_{T_2}(t))\\
\ \ \ \ \ \ \ \ \ = (1-(1-e^{-\int_{0}^{t} h_1(s)\ ds}))(1-(1-e^{-\int_{0}^{t} h_2(s)\ ds}))\\
\ \ \ \ \ \ \ \ \ =e^{-\int_{0}^{t} (h_1(s)+h_2(s))\ ds}\\
\ \ \ \ \Rightarrow F_T(t) = 1 - e^{-\int_{0}^{t} \underbrace{(h_1(s)+h_2(s))}_{\text{failure rate function of $T$}}\ ds}
$

$
\textbf{Binary hypothesis testing with continuous-type observations}\\
\ \\
\text{Two hypotheses:}\\
\ \ \ \ \text{$H_0$: $X$ has pdf $f_0$}\\
\ \ \ \ \text{$H_1$: $X$ has pdf $f_1$}\\
\text{Given $X=u$, which $H_i$ do we choose?}\\
\text{Likelihood ratio}\ \Lambda(u)\ = \frac{f_1(u)}{f_0(u)}\ \begin{cases}
> \tau & \text{choose $H_1$}\\
< \tau & \text{choose $H_0$}\\
\end{cases}\\
\textbf{ML decision rule:}\ \text{Choose $H_i$ for which $f_i(u)$ is larger}\\
\ \ \ \ \text{This is equivalent to}\ \Lambda(u)\ \begin{cases}
> 1 & \text{choose $H_1$}\\
< 1 & \text{choose $H_0$}\\
\end{cases}\\
\textbf{MAP decision rule:}\ \text{Assume prior probabilites are known}\\
\ \ \ \ \text{$P(H_0)=\pi_0$ and $P(H_1)=\pi_1=1-\pi_0$}\\
\ \ \ \ \text{Choose $H_i$ for which $\pi_i f_i(u)$ is larger}\\
\ \ \ \ \text{Equivalent to}\ \Lambda(u)\ \begin{cases}
> \frac{\pi_0}{\pi_1} & \text{choose $H_1$}\\
< \frac{\pi_0}{\pi_1} & \text{choose $H_0$}\\
\end{cases}\\
\textbf{Error Probabilities:}\ \text{same as discrete-type case}\\
\ \ \ \ P_{\text{false alarm}} = P(\text{choose}\ H_1|H_0),\ P_{\text{miss}} = P(\text{choose}\ H_0|H_1)\\
\ \ \ \ P_e = \pi_0 P_{\text{false alarm}} + \pi_1 P_{\text{miss}}\\
\ \ \ \ \text{As in discrete-type case, MAP rule minimizes $P_e$}\\
$

$
\text{pdfs:}\ f_i(u) = \frac{1}{\sqrt{2\pi \sigma^2}} exp(-\frac{(u-m_i)^2}{2\sigma^2})\\
\textbf{ML decision rule:}\\
\ \ \ \ \Lambda (u) = \frac{f_1(u)}{f_0(u)} = \frac{exp(-\frac{(u-m_1)^2}{2\sigma^2})}{exp(-\frac{(u-m_0)^2}{2\sigma^2})}
= exp (\frac{-(u-m_1)^2+(u-m_0)^2}{2\sigma^2})\\
\ \ \ \ = exp (-\frac{1}{2\sigma^2} (u^2 - 2um_1 + m_1^2 - u^2 + 2um_0 - m_0^2))\\
\ \ \ \ = exp (\frac{1}{2\sigma^2} (2u(m_1-m_0) - (m_1+m_0)(m_1-m_0)))\\
\ \ \ \ = exp (\underbrace{\frac{1}{2\sigma^2}}_{>0} \underbrace{(m_1-m_0)}_{>0}(2u-(m_1+m_0)))\ \begin{cases}
> 1 & \text{choose $H_1$}\\
< 1 & \text{choose $H_0$}\\
\end{cases}\\
\ \ \ \ \text{Equivalent to}\ 2u - (m_1+m_0)\ \begin{cases}
> 0 & \text{choose $H_1$}\\
< 0 & \text{choose $H_0$}\\
\end{cases}\\
\ \ \ \ \Leftrightarrow X \begin{cases}
> \frac{m_1+m_0}{2} & \text{choose $H_1$}\\
< \frac{m_1+m_0}{2} & \text{choose $H_0$}\\
\end{cases}\ \text{($\frac{m_1+m_0}{2}$ is the mid-point, so make sense!)}
$


$
\textbf{MAP decision rule:}\ \Lambda(u)\ \begin{cases}
> \frac{\pi_0}{\pi_1} & \text{choose $H_1$}\\
< \frac{\pi_0}{\pi_1} & \text{choose $H_0$}\\
\end{cases}\\
\ \ \ \ \Lambda(u) = exp \{\frac{1}{2\sigma^2} (m_1-m_0) (2u-(m_1+m_0))\} \gtrless \frac{\pi_0}{\pi_1}\\
\ \ \ \ \Rightarrow \frac{1}{2\sigma^2} (m_1-m_0) (2u-(m_1+m_0)) \gtrless \ln \frac{\pi_0}{\pi_1}\\
\ \ \ \ u \gtrless \underbrace{\frac{\sigma^2}{m_1-m_0} \ln \frac{\pi_0}{\pi_1} + \frac{m_1+m_0}{2}}_{\gamma_{MAP}}\\
\ \ \ \ X \begin{cases}
> \gamma_{MAP} & \text{choose $H_1$}\\
< \gamma_{MAP} & \text{choose $H_0$}\\
\end{cases}\\
$


$
\text{Both ML and MAP rules are expressed as}\ \begin{cases}
> \gamma & \text{choose $H_1$}\\
< \gamma & \text{choose $H_0$}\\
\end{cases}\\
P_{\text{false alarm}} = P(X > \gamma | \underbrace{H_0}_{\text{$X$ is $N(m_0,\sigma^2)$}}) 
= P(\frac{X-m_0}{\underbrace{\sigma}_{N(0,1)}} > \frac{\gamma-m_0}{\sigma}|H_0) = Q(\frac{\gamma-m_0}{\sigma})\\
P_{\text{miss}} = P(X<\gamma | H_1) = Q(\frac{m_1-\gamma}{\sigma})\\
P_e = \pi_0 P_{\text{false alarm}} + \pi_1 P_{\text{miss}}
$
