\textbf{Previously}\\
\ \\
\bullet \text{CDF applies to all RVs}\\
\ \ \ \ -\ \text{For a RV $X$, CDF is}\ F_X(c) = P\{X\le c\}\\
\ \ \ \ -\underbrace{F_X(c-)}_{\text{left-limit}} = P\{X<c\}\\
\ \ \ \ -\Delta F_X(c) = \underbrace{F_X(c)-F_X(c-)}_{\text{size of jump at $c$}} = P\{X=c\}\\
$

$
\textbf{Ex 3.1.4}\ \text{Let $X$ have the following CDF}\\
\ \ \ \ \text{a)}\ P\{X\le 1\} = F_X(1) = \text{slope}\cdot 1 = \frac{0.25}{5}\cdot 1 = 0.05\\
\ \ \ \ \text{b)}\ P\{X\le 10\} = F_X(10) = 0.75\\
\ \ \ \ \text{c)}\ P\{X\ge 10\} = 1-P\{x<10\} = 1-\overbrace{F_X(10-)}^{0.5} = 0.5\\
\ \ \ \ \text{d)}\ P\{X=10\} = \Delta F_X(10) = 0.25\\
\ \ \ \ \text{e)}\ P\{|X-5| \le 0.1\} = P\{4.9 \le X \le 5.1\} = P\{X\le 5.1\}-P\{X<4.9\} 
= F_X(5.1) - F_X(4.9-) = 0.5 - \frac{0.25}{5}\times 4.9
$


$
\textbf{Properties of the CDF}\ \text{(necessary and sufficient)}\\
\ \ \ \ \text{1) $F_X$ is non-decreasing}\\
\ \ \ \ \text{2) $\lim_{c\to \infty} F_X(c)=1$, $\lim_{c\to -\infty} F_X(c)=0$}\\
\ \ \ \ \text{3) $F_X$ is right-continuous, i.e. $F_X(c)=F_X(c+)$ for all c}
$

$
\textbf{Continuous-type RV}\ \text{(3.2)}\\
\ \\
\bullet \text{Recall: a discrete-type RV $X$ takes value in finite or countably infinite set of values}\ \{U_1,\ U_2,\ U_3,\ ...\}\\
\bullet \text{A RV $X$ is continuous-type if there exists a function}\\
\ \ \ f_X\ \text{(called probability density function) such that}\ F_X(c) = \int_{-\infty}^{c} f_X(u)\ du\\
$


$
\bullet\ \text{From calculus,}\ F_X'(c) = f_X(c)\ \text{(but CDF does not have to be differentiable everywhere)}\\
\ \ \ \text{(this only holds if $F_X$ is}\ \textbf{differentiable}\ \text{at $c$)}
$


$
\bullet\ \text{Because CDF is non-decreasing, pdf $\ge$ 0}\\
\bullet \text{CDF of continuous-type RV is continuous $\rightarrow$ no jumps. Therefore,}\ P\{X=c\} = 0\ \text{for all $c$}\\
\bullet \text{For a $<$ b,}\ P\{a<X\le b\} = F_X(b) - F_X(a) = \int_{-\infty}^b f_X(u)\ du - \int_{-\infty}^a f_X(u)\ du = \int_a^b f_X(u)\ du\\
\bullet \text{Since}\ P\{X=a\} = P\{X=b\} = 0,\\
\ \ \ P\{a<X\le b\} = P\{a\le X< b\} = P\{a\le X\le b\} = P\{a<X<b\} = \int_a^b f_X(u)\ du,\ \text{for continuous-type RVs}\\
\bullet \int_{-\infty}^{\infty} f_X(u)\ du = 1\\
\ \\
\textbf{Ex 3.2.2}\ \text{Suppose $X$ has pdf}\ f_X(u) = 
\begin{cases}
A(1-u^2), & -1\le u \le 1\\
0, & \text{otherwise}
\end{cases}\\
\ \ \ \ \text{a) Find $A$. Using}\ \int_{-\infty}^{\infty} f_X(u)\ du = 1,\ \text{we found}\ A=\frac{3}{4}\\
\ \ \ \ \text{b)}\ P\{0.5<X<1.5\} = \int_{0.5}^{1.5} f_X(u)\ du = \int_{0.5}^1 \frac{3}{4}(1-u^2)\ du\\
\ \ \ \ \text{c) Find CDF}\ F_X(c) = \int_{-\infty}^c f_X(u)\ du\\
\ \ \ \ \ \ \ \ F_X(c) = \begin{cases}
0, & c < -1\\
\int_{-1}^c \frac{3}{4} (1-u^2)\ du = \frac{3c}{4} - \frac{c^3}{4} +\frac{1}{2}, & -1 \le c \le 1\\
1, & c > 1
\end{cases}\\
\ \ \ \ \text{d)}\ \mu_X = E[X] = \int_{-\infty}^{\infty} u f_X(u)\ du = 0,\ \text{because pdf is symmetric ($f_X(u)=f_X(-u)$)}\\
\ \ \ \ \ \ \ \ \text{similarly,}\ E[X^3] = E[X^5]=\ ...\ = 0\ \text{(check)}\\
\ \ \ \ \text{e) Var}(X)=E[X^2] - E[X]^2 = \int_{-\infty}^{\infty} u^2 f_X(u)\ du = \frac{1}{5}
$
