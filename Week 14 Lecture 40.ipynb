\textbf{Previously}\ \text{Joint Gaussian distribution (bivariate normal distribution)}\\
\ \\
\textbf{Definition:}\ \text{RVs $X$ and $Y$ are jointly Gaussian if $aX+bY$ is Gaussian for all $a,b\in \mathbb{R}$}\\
\ \\
\text{Note: by setting $a=1,b=0$, $X$ must be Gaussian, and $b=1,a=0$, $Y$ must be Gaussian}\\
\text{As long as $\rho = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}\neq 1$ and $\rho \neq -1$, $X$ and $Y$ have a joint pdf,}\\
\text{given by}\ f_{XY}(u,v) = \frac{1}{2\pi \sigma_X \sigma_Y \sqrt{1-\rho^2}} 
exp(-\frac{(\frac{u-\mu_X}{\sigma_X})^2 + (\frac{v-\mu_Y}{\sigma_Y})^2 - 2\rho (\frac{u-\mu_X}{\sigma_X})(\frac{v-\mu_Y}{\sigma_Y})}{2(1-\rho^2)})\\
\ \\
\text{special case: standard bivariate normal distribution $X$ and $Y$ are both $N(0,1)$ RVs and they are independent}\\
f_{XY}(u,v) = f_X(u)f_Y(v) = \frac{1}{\sqrt{2\pi}} e^{-\frac{u^2}{2}}\cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{v^2}{2}}
= \frac{1}{2\pi} e^{-\frac{u^2+v^2}{2}}\\
\text{(special case where $\rho=0,\ \mu_X=\mu_Y=0,\ {\sigma_X}^2 = {\sigma_Y}^2=1$)}
$

$
\rho = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y} = \frac{-1}{\sqrt{5}\sqrt{2}} = \frac{-1}{\sqrt{10}} \neq 1\ \text{or} \ -1\
\implies \text{joint pdf $f_{XY}(u,v)$,}\\
P(X+2Y\ge 1) = P(Y\ge \frac{1-X}{2})
= \int_{-\infty}^{\infty} \int_{\frac{1-u}{2}}^{\infty}\ f_{XY}(u,v)\ dvdu \to \text{hard!}\\
\text{Easier way: by definition of jointly Gaussian, $X+2Y$ is also Gaussian.}\ E[X+2Y] = E[X]+2E[Y] = 0\\
\mbox{Var}(X+2Y) = \mbox{Cov}(X+2Y,X+2Y) = \mbox{Cov}(X,X) + 2\mbox{Cov}(X,2Y) + \mbox{Cov}(2Y,2Y)\\
\ \ \ \ = {\sigma_X}^2 + 4\cdot \mbox{Cov}(X,Y) + 4{\sigma_Y}^2 = 5 + 4\times (-1) + 8 = 9\\
\implies X+2Y\ \text{has a $N(0,9)$ distribution}\\
P(X+2Y\ge 1) = P(\frac{X+2Y}{\sqrt{9}}\ge \frac{1}{\sqrt{9}}) = P(\overbrace{\frac{X+2Y}{\sqrt{9}}}^{N(0,1)}\ge \frac{1}{3})
= Q(\frac{1}{3}) \approx 0.3694
$

$
\textbf{Key properties of the bivariate normal distribution}\\
\textbf{Proposition}\ \text{Suppose $X,Y$ have a bivariate normal distribution, with parameters $\mu_X, \mu_Y, {\sigma_X}^2, {\sigma_Y}^2, \rho$. Then}\\
\text{a) $X$ has $N(\mu_X,{\sigma_X}^2)$ and $Y$ has $N(\mu_Y,{\sigma_Y}^2)$}\\
\text{b) $aX+bY$ is Gaussian for any $a,b\in \mathbb{R}$}\\
\text{c) $\rho$ is the correlation coefficient between $X$ and $Y$}\\
\text{d) $X$ and $Y$ are independent if and only if $\rho=0$}\\
\ \ \ \ \text{(if two jointly Gaussian RVs are uncorrelated, they are also independent)}\\
\text{e) The unconstrained MMSE estimator of $Y$ given $X$ is linear:}\
E[Y|X] = \hat{E}[Y|X]\ (\mbox{i.e.}\ g^*(X) = L^*(X))\\
\text{f) The conditional distribution of $Y$ given $X=u$ is}\\
\ \ \ \ N(\hat{E}[Y|X=u], {\sigma_e}^2),\ \text{where}\ {\sigma_e}^2 
= \underbrace{E[(Y-\hat{E}[Y|X])^2]}_{\text{MSE of estimating $Y$ from $X$}}
$

$
\text{Let's prove some of these facts.}\\
\text{We focus on the special case $\mu_X=\mu_Y=0$, ${\sigma_X}^2={\sigma_Y}^2=1$. Then}\\
f_{XY}(u,v) = \frac{1}{2\pi \sqrt{1-\rho^2}} exp(-\frac{u^2+v^2-2\rho u v}{2(1-\rho^2)})\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \  
= \frac{1}{2\pi \sqrt{1-\rho^2}} exp (-\frac{u^2(1-\rho^2) + \overbrace{u^2\rho^2 + v^2 - 2\rho u v}^{(v-\rho u)^2}}{2(1-\rho^2)})\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \  
= \underbrace{[\frac{1}{\sqrt{2\pi}} exp(-\frac{u^2}{2})]}_{A(u)} \cdot
\underbrace{[\frac{1}{\sqrt{2\pi(1-\rho^2)}} exp(-\frac{(v-\rho u)^2}{2(1-\rho^2)})]}_{B(u,v)}\\
f_X(u) = \int_{-\infty}^{\infty}\ f_{XY}(u,v)\ dv = A(u) \underbrace{\int_{-\infty}^{\infty} B(u,v)\ dv}_{=1}
= \frac{1}{\sqrt{2\pi}} exp(-\frac{u^2}{2}) \to \text{$X$ is $N(0,1)$}\\
f_{Y|X}(v|u) = \frac{f_{XY}(u,v)}{f_X(u)} = \frac{A(u)B(u,v)}{A(u)} 
= \frac{1}{\sqrt{2\pi (1-\rho^2)}} exp(-\frac{(v-\rho u)^2}{2(1-\rho^2)})\\
\text{For a fixed $u$, this is the pdf of $N(\rho u, 1-\rho^2)$}\\
\text{In order for $X$ and $Y$ to be independent, we need $f_{Y|X}(v|u) = f_Y(v)$,}\\
\text{i.e., $\frac{1}{\sqrt{2\pi (1-\rho^2)}} exp(-\frac{(v-\rho u)^2}{2(1-\rho^2)}) = \frac{1}{\sqrt{2\pi}} exp(-\frac{v^2}{2})$ for all $v$.}\\
\text{This happens if and only if $\rho=0$. (this proves (d))}\\
\text{Estimating $Y$ from $X=u$:}\\
\text{unconstrained estimator:}\\
E[Y|X=u] = \int_{-\infty}^{\infty}\ v f_{Y|X}(v|u)\ dv = \rho u\ \ 
(\underbrace{\text{$Y$ is $N(\rho u, 1-\rho^2)$}}_{\text{given $X=u$}})\\
\text{Linear estimator:}\\
\hat{E}[Y|X=u] = \underbrace{E[Y]}_{=0} + \frac{\overbrace{\text{Cov}(X,Y)}^{\rho}}{\underbrace{\text{Var}(X)}_{=1}}
(X-\underbrace{E[X]}_{=0}),\ \rho = \frac{\text{Cov}(X,Y)}{\underbrace{\sigma_X}_{=1} \underbrace{\sigma_Y}_{=1}}\\
\hat{E}[Y|X=u] = \rho u = E[Y|X=u]\ \ (\text{this proves (e)})\\
\underbrace{\text{MSE}}_{{\sigma_e}^2} = E[(Y-\hat{E}[Y|X])^2] 
= \underbrace{\mbox{Var}(Y)}_{=1} - \frac{\text{Cov}(X,Y)^2}{\underbrace{\text{Var}(X)}_{=1}} = 1-\rho^2\\
\text{Given $X=u$, $Y$ is $N(\rho u, 1-\rho^2) = N(\hat{E}[Y|X=u], {\sigma_e}^2)$. This proves (f).}
$

$
\text{Given $X=u$, $Y$ is distributed as $N(\rho u, 1-\rho^2)$}\\
\text{I can think of a new RV $Z$ with distribution $N(\rho u, 1-\rho^2)$}\\
E[Y^2|X=u] = E[Z^2]\\
\mbox{Var}(Z) = 1-\rho^2 = E[Z^2] - E[Z]^2 = E[Z^2] - \rho^2 u^2\\
\implies E[Y^2|X=u] = E[Z^2] = 1-\rho^2 + \rho^2 u^2\\
\implies E[Y^2|X] = 1-\rho^2 + \rho^2 X^2
$
