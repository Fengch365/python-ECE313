$
\textbf{Law of large numbers (LLN) and Central Limit Theorem (CLT)}\\
\textbf{Idea:}\
\text{Suppose $X_1,\ X_2,\ ...$ are RVs with mean $\mu$ and $S_n=\sum_{i=1}^{n} X_i$. Then $\frac{S_n}{n} \approx \mu$ for large $n$.}\\
\textbf{Note: $\frac{S_n}{n}$ is a RV while $\mu$ is a constant}\\
\ \\
\text{There are many versions of LLN. We consider the following:}\\
\textbf{Proposition:}\ \text{Suppose $X_1,\ X_2,\ ...$ are uncorrelated RVs and}\\
\ \ \ \ \bullet E[X_i] = \mu\ \mbox{for}\ i=1,2,3,...\\
\ \ \ \ \bullet \mbox{Var}(X_i) \le C\ \mbox{for}\ i=1,2,3,...\\
\ \ \ \ \text{Note: $X_1,\ X_2,\ ...$ are not assumed to be independent}\\
\text{Then, for any $\delta > 0$,}\ P\{|\frac{S_n}{n}-\mu|\ge \delta\} \le \frac{C}{n\delta^2} \to 0\ \mbox{as}\ n\to \infty.\\
\textbf{proof:}\ \text{we will use Chebyshev's inequality:}\ P\{|Y-E[Y]|\ge d\} \le \frac{\text{Var}(Y)}{d^2}\ \text{for every $d$}\\
\text{Pick $Y=\frac{S_n}{n}$}\\
\bullet E[\frac{S_n}{n}] = \frac{1}{n} \sum_{i=1}^{n} \underbrace{E[X_i]}_{=\mu} = \mu\\
\bullet \mbox{Var}(\frac{S_n}{n}) = \frac{1}{n^2} \mbox{Var}(\sum_{i=1}^{n}X_i) \underbrace{=}_{\text{$S_i$'s are uncorrelated}}
\frac{1}{n^2} \sum_{i=1}^{n} \mbox{Var}(X_i) \le \frac{1}{n^2} nC = \frac{C}{n}\\
\text{By Chebyshev's,}\ P\{|\frac{S_n}{n}-\mu|\ge \delta\} \le \frac{\text{Var}(\frac{S_n}{n})}{\delta^2} \le \frac{C}{n\delta^2}\\
\ \\
\textbf{Ex.}\ \text{$X_1,\ X_2,\ ...$ are iid (independent and identically distributed) and uniformly distributed in $[0,1]$}\\
\ \ \ \ E[X_i] = \frac{1}{2},\ \mbox{Var}(X_i) = \frac{1}{12}\\
\text{For large $n$, LLN implies that}\ P\{|\frac{S_n}{n}-\frac{1}{2}|\ge \delta\} \ge \frac{1/12}{n\delta^2}\\
\text{Intuitively, for large $n$, $\frac{S_n}{n} \approx \frac{1}{2}$, $S_n\approx \frac{n}{2}$}
$

$
\text{Let $X_i\in \{1,2,3,...,6\}$ be the outcome of $i^{th}$ roll}\\
S_{1000} = \sum_{i=1}^{1000},\ E[X_i] = \frac{1+2+3+4+5+6}{6} = 3.5\\
\text{By LLN,}\ \frac{S_{1000}}{1000} \approx 3.5 \implies S_{2000} \approx 3500
$

$
\text{Note: $X_1,\ X_2,\ ...,\ X_{100}$ are not uncorrelated. Cannot use LLN.}\\
\ \ \ \ \mbox{Cov}(X_i,X_i) = \mbox{Var}(X_i) = 1,\ \mbox{Cov}(X_i,X_{i+1}) \le 0.1\\
\ \ \ \ \mbox{Cov}(X_i,X_{i+2}) = \mbox{Cov}(X_i,X_{i+3}) = ... = 0\\
\text{a) Var}(S_{100}) = \mbox{Var}(\sum_{i=1}^{100} X_i) = \sum_{i=1}^{100}\sum_{j=1}^{100} \mbox{Cov}(X_i,X_j)\\
\ \ \ \ = \sum_{i=1}^{100} \mbox{Var}(X_i) + \sum_{i=1}^{99} \mbox{Cov}(X_i,X_{i+1}) + \sum_{i=2}^{100} \mbox{Cov}(X_i,X_{i-1})\\
\ \ \ \ \le 100\cdot 1 + 99\cdot 0.1 + 99\cdot 0.1 = 118 < 120\\
\text{b)}\ E[\frac{S_{100}}{100}] = \frac{1}{100}\cdot \sum_{i=1}^{100} E[X_i] = 5.\\
\ \ \ \ \text{By Chebyshev's:}\ P\{|\frac{S_{100}}{100}-5|\ge 0.5\} \le \frac{\text{Var}(\frac{S_{100}}{100})}{0.5^2} = \frac{\text{Var}(S_{100})}{100^2 \times 0.5^2} \le \frac{120}{100^2 \times 0.5^2} = 0.048\\
\text{Note: This problem shows that one can derive a LLN for settings other than the one in the proposition}
$

$
\text{a) Need to find $\lambda$ that maximizes (Joint) likelihood}\\
\ \ \ \ f_{U_1U_2...U_n}(u_1,u_2,...u_n) = f_{U_1}(u_1)...f_{U_n}(u_n) = \prod_{i=1}^{n} f_{U_i}(u_i)\ 
=\prod_{i=1}^{n} \lambda e^{-\lambda u_i} = \lambda^n e^{-\lambda \sum_{i=1}^{n}u_i}\\
\ \ \ \ \frac{d}{d\lambda} (\lambda^n e^{-\lambda \sum u_i}) 
= n\lambda^{n-1} e^{-\lambda \sum u_i} + \lambda^n (-\sum_{i=1}^{n}u_i) e^{-\lambda \sum u_i}\
= \underbrace{\lambda^{n-1} e^{-\lambda \sum u_i}}_{>0} (n-\lambda \sum_{i=1}^{n} u_i) = 0\\
\ \ \ \ \implies \hat{\lambda}_{ML} = \frac{n}{\sum_{i=1}^{n}u_i} = \frac{n}{S_n}\\
\text{b) We want}\ P\{\lambda \in [0.9\hat{\lambda}_{ML}, 1.1\hat{\lambda}_{ML}]\} \ge 0.96\\
\ \ \ \ P\{|\lambda-\hat{\lambda}_{ML}|\le 0.1 \hat{\lambda}_{ML}\} = P\{|\frac{\lambda}{\hat{\lambda}_{ML}}|\le 0.1\} \ge 0.96\\
\ \ \ \ \iff P\{|\frac{\lambda}{\hat{\lambda}_{ML}}-1| > 0.1\} \le 0.04\ \text{(what I want)}\\
\ \ \ \ E[\frac{\lambda}{\hat{\lambda}_{ML}}] = \lambda E[\frac{S_n}{n}] = \lambda \cdot E[X_1] = \lambda \cdot \frac{1}{\lambda} = 1\\
\ \ \ \ \mbox{Var}(\frac{\lambda}{\hat{\lambda}_{ML}}) = \mbox{Var}(\frac{\lambda S_n}{n}) = \frac{\lambda^2}{n^2}\mbox{Var}(S_n)
= \frac{\lambda^2}{n^2} \cdot n \cdot \frac{1}{\lambda^2} = \frac{1}{n}\\
\ \ \ \ \text{By Chebyshev's:}\ P\{|\frac{\lambda}{\hat{\lambda}_{ML}}-1|>0.1\} 
\le \frac{\text{Var}(\frac{\lambda}{\hat{\lambda}_{ML}})}{0.1^2} = \frac{100}{n}\\
\ \ \ \ \text{I want}\ \frac{100}{n} = \frac{1}{25} \implies n = 2500

$
\textbf{Central Limit Theorem}\\
\textbf{Recall:}\
\text{DeMoivre-Laplace theorem: If $X_1,\ X_2,\ ...$ are iid Bernoulli($p$), then}\ \lim_{n\to \infty} P\{\frac{S_n-np}{\sqrt{np(1-p)}} \le c\} = \Phi(c)\\
\text{In words: for large $n$, $\frac{S_n-np}{\sqrt{np(1-p)}}$ is approximately $N(0,1)$}\\
\text{The CLT generalizes this result to non-Bernoulli RVs}\\
\textbf{Theorem (CLT):}\ \text{Suppose $X_1,\ X_2,\ ...$ are iid RVs with mean $\mu$ and variance $\sigma^2$ (assumed finite).}\\
\text{Then}\ \lim_{n\to \infty} P\{\frac{S_n-n\mu}{\sqrt{n\sigma^2}}\le c\} = \Phi(c)\\
\text{This allows us to use the Gaussian approximation for sums of iid non-Bernoulli RVs.}
$

$
X_i\in \{1,2,...,6\},\ S = \sum_{i=1}^{1000} X_i\\
E[S] = 1000 \cdot E[X_1] = 3500\\
\mbox{Var}(S) = 1000\cdot \mbox{Var}(X_1) = 1000 (E[{X_1}^2] - E[X_1]^2) = 2916.7\\
\text{From CLT,}\ P\{\frac{S-3500}{\sqrt{2916.7}} \le c\} \approx \Phi(c)\\
\implies P\{|S-3500| \le L\} = P\{|S-3500| \le L+\frac{1}{2}\} 
= P\{|\underbrace{\frac{S-3500}{54.01}}_{\approx N(0,1)}|\le \underbrace{\frac{L+\frac{1}{2}}{54.01}}_{A}\}
\approx 1 - 2Q(\frac{L+\frac{1}{2}}{54.01}) \approx 0.9 \implies L \approx 88
$
