$
\textbf{Cumulative Distribution Function (CDF)}\\
\ \\
\text{CDF of $X$ is}\ \textbf{defined}\ \text{as}\ F_X(c) = P\{X\le c\}\\
\ \\
\bullet \text{All RVs have a CDF}\\
\bullet \text{CDF is defined for all real values}\\
\bullet \text{CDF does not need to be continuous}\\
\ \\
\textbf{Ex 3.1.1}\ \text{You roll a die. $X=$ number shown.}\\
X\in \{1,2,3,4,5,6\},\ P_X(i) = \frac{1}{6}\ \text{for}\ i=1,2,3,4,5,6\\
\ \ \ \ F_X(x) = P\{X\le x\}
$

$
\text{Useful notation}\\
\bullet \text{$F_X(x-)$ is the left limit of $F_X$ at $x$}\\
\bullet \text{$F_X(x+)$ is the right limit of $F_X$ at $x$}\\
\ \\
\text{For previous example,}\\
\ \ \ \ F_X(2-) = \frac{1}{6}\\
\ \ \ \ F_X(2+) = \frac{2}{6}\\
\ \\
\text{Notice that}\ F_X(c-) = P\{X<c\}\\
\text{Hence,}\ \underbrace{F_X(c) - F_X(c-)}_{\Delta F_X(c)\rightarrow \text{size of the jump}} = P\{X\le c\} - P\{X<c\} = P\{X=c\}\\
\ \\
\text{For previous example,}\ \Delta F_X(2.5) = 0 = P\{X=2.5\}
$

$
\textbf{Ex 3.1.3}\ \text{CDF of $X$ is $F_X$}\\
\ \ \ \ \text{a) what values of $u$, is}\ P\{X=u\}>0\ ?\\
\ \ \ \ \ \ \ \ \text{For $u=0$, $\Delta F_X(0) =0.5 = P\{X=0\}$}\\
\ \ \ \ \ \ \ \ \text{No other jumps}\\
\ \ \ \ \text{b) Find}\ P\{X\le 0\} = F_X(0) =1\\
\ \ \ \ \text{c) Find}\ P\{X< 0\} = F_X(0-) =0.5
$


$
\textbf{Properties of the CDF}\\
\ \ \ \ \text{1) $F_X$ is non-decreasing}\\
\ \ \ \ \text{2) $\lim_{c\to \infty} F_X(c)=1$, $\lim_{c\to -\infty} F_X(c)=0$}\\
\ \ \ \ \text{3) $F_X$ is right-continuous, i.e. $F_X(c)=F_X(c+)$ for all c}
$
