$
\textbf{Previously:}\ \text{Jointly distributed RVs}\\
\text{(RVs $X$ and $Y$ defined on the same probability space)}\\
\bullet \text{Joint CDF of $X$ and $Y$}\\
\ \ \ F_{X,Y}(u,v) = P(X\le u,Y\le v)
$

$
\bullet\ \text{If $X$ and $Y$ are discrete-type RVs, they have a joint pmf.}\\
\ \ \ P_{X,Y}(u,v) = P(X=u,Y=v)\\
\ \ \ \text{$X$ takes value in $\{u_1,u_2,...\}$, $Y$ takes value in $\{v_1,v_2,...\}$}\\
\ \ \ \sum_i \sum_j p_{X,Y}(u_i,v_j) = 1\\
\ \ \ \textbf{marginalization}:\\
\ \ \ p_X(u) = \sum_j p_{X,Y}(u,v_j)\\
\ \ \ p_Y(v) = \sum_i p_{X,Y}(u_i,v)\\
\ \ \ \textbf{conditional pmf}:\\
\ \ \ P_{Y|X}(v|u) = P(Y=v|X=u) = \frac{P(Y=v,X=u)}{P(X=u)} = \frac{p_{X,Y}(u,v)}{p_X(u)}\ \text{if $p_X(u)>0$ (undefined otherwise)}
$


$
\text{a)}\ p_X(u) = \sum_j p_{X,Y} (u,v_j) = p_{X,Y} (u,1) + p_{X,Y} (u,2) + p_{X,Y} (u,3)\\
p_X(1) = 0.1,\ p_X(2) = 0.3+0.2+0.1 = 0.6,\ p_X(3) = 0.2+0.1 = 0.3\\
\text{b)}\ p_Y(v) = \sum_i p_{X,Y} (u_i,v) = p_{X,Y} (1,v) + p_{X,Y} (2,v) + p_{X,Y} (3,v)\\
p_Y(1) = 0.4,\ p_Y(2) = 0.4,\ p_Y(3) = 0.2\\
\text{c)}\ P(X=Y) = p_{X,Y} (1,1) + p_{X,Y} (2,2) + p_{X,Y} (3,3) = 0+0.2+0=0.2\\
\text{d)}\ P(X>Y) = 0.3+0.2+0.1 = 0.6\\
\text{e)}\ p_{Y|X}(v|2) = P(Y=v|X=2) = \frac{P(Y=v,X=2)}{P(X=2)} = \frac{p_{X,Y}(2,v)}{p_X(2)} = \frac{p_{X,Y}(2,v)}{0.6}
$

$
\textbf{Joint Probability Density Function (pdf)}\\
\text{RVs $X$ and $Y$ are called jointly continuous-type if there exists a function $f_{X,Y}$ (the joint pdf)}\\
\text{such that, for any $(u_0,v_0)\in \mathbb{R}^2$,}\ 
F_{X,Y}(u_0,v_0) = \int_{-\infty}^{u_0} \int_{-\infty}^{v_0} f_{X,Y}(u,v)\ dvdu
$


$
\text{In general, we integrade on a region $R$ to get $P((X,Y)\in R)$}
$

$
\text{Properties of the joint pdf:}\\
\text{1) For any $(u,v)\in \mathbb{R}^2$, $f_{X,Y}(u,v)\ge 0$}\\
\text{2)}\ \underbrace{\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dudv = 1}_{P((X,Y)\in \mathbb{R}^2)=1}
$

$
\textbf{LOTUS}\\
\text{If $g(x,y)$ is a function on $\mathbb{R}^2$, we have}\ 
E[g(X,Y)] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} g(x,y) f_{X,Y}(x,y)\ dxdy\\
\text{LOTUS can be used to prove}\ \textbf{linearity of expectation}\Rightarrow E[aX+bY+c] = aE[X]+bE[Y]+c\\
\textbf{Proof}\ \text{Let}\ g(x,y) = ax+by+c.\ \text{Then}\\
E[aX+bY+c] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (ax+by+c) f_{X,Y}(x,y)\ dxdy\\
= a\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} x f_{X,Y}(x,y)\ dxdy 
+ b\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} y f_{X,Y}(x,y)\ dxdy 
+ c\underbrace{\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(x,y)\ dxdy}_{1}\\
= aE[X] + bE[Y] + c\\
\ \\
\textbf{marginal pdfs}\\
\text{From $f_{X,Y}$ we can compute the marginal pdfs $f_x$ and $f_y$}\\
F_X(u_0) = P(X\le u_0) = P(X\le u_0,Y\le \infty) = \int_{-\infty}^{u_0} \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dudv
= \int_{-\infty}^{u_0} [\int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dv]\ du = \int_{-\infty}^{u_0} [f_X(u)]\ du\\
\text{We see that}\ f_X(u) = \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dv\ \text{(marginalization)}\\
\text{Similarily,}\ f_Y(v) = \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ du\\
\textbf{Note} \text{: this is similar to marginalization for joint pmfs}\Rightarrow p_X(u) = \sum_j p_{X,Y}(u,v_j)\\
\ \\
\textbf{Conditional pdf}\ \text{of $Y$ given $X$ is given by}\\
f_{Y|X} (v|u_0) = \underbrace{\frac{f_{X,Y}(u_0,v)}{f_X(u_0)}}_{\text{a pdf in $v$}}\ \text{if $f_X(u_0)>0$ (undefined otherwise)}
$

$
\text{We divide by $f_X(u_0)$ so that $f_{Y|X}$ integrates to 1}\\
\int_{-\infty}^{\infty} f_{Y|X}(v|u_0)\ dv = \int_{-\infty}^{\infty} \frac{f_{X,Y}(u_0,v)}{f_X(u_0)}\ dv
= \frac{1}{f_X(u_0)} \underbrace{\int_{-\infty}^{\infty} f_{X,Y}(u_0,v)\ dv}_{\text{marginalization}\to f_X(u_0)}
= \frac{1}{f_X(u_0)} \cdot f_X(u_0) = 1\\
\text{Conditional pdf $f_{Y|X}(v|u_0)$, for a fixed $u_0$, is another pdf in $v$}\\
\ \\
\textbf{Conditional Expectation}\\
E[Y|X=u] = \int_{-\infty}^{\infty} y\cdot f_{Y|X}(y|u)\ dy = g(u)\Rightarrow \text{a deterministic number}\\
\textbf{Note}\text{: this is a function of $u$}\\
\text{We define $\underbrace{E[Y|X]}_{\text{random variable}}=g(X)$}
$


$
\text{a) $f_{X,Y}$ must integrate to 1}\\
1 = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dudv = \int_{u=0}^{1} \int_{v=0}^{1-u} c(1-u-v)\ dvdu\\
= c \int_0^1 (v-uv-\frac{v^2}{2})_0^{1-u}\ du = c \int_0^1 (\overbrace{1-u-u+u^2}^{(1-u)^2} - \frac{(1-u)^2}{2})\ du\\
= c \int_0^1 \frac{(1-u)^2}{2}\ du = \frac{c}{2} \int_0^1 (u^2-2u+1)\ du\\
= \frac{c}{2} (\frac{u^3}{3}-u^2+u)_0^1 = \frac{c}{6} = 1 \Rightarrow c=6
$
