\textbf{Reliability}\ \text{(compute/bound for failure probability)}\\
\ \\
\textbf{Union bound}\ \text{For events}\ A_1,\ A_2,\ A_3,\ ...,\ A_m\\
\ \ \ \ P(A_1\cup A_2\cup\ ...\ \cup A_m) \le P(A_1) +\ ...\ + P(A_m)\\
\ \ \ \ \text{(equality if events are}\ \textbf{mutually exclusive}\text{)}\\
\ \ \ \ \text{(mainly useful when $P(A_i)$'s are small)}\\
\ \\
\textbf{SAQ 2.12}\ \text{Storage system with 5 servers. System fails if 3 or more crash.}\\
\text{Each one crashes independently with probability $p=0.001$. Bound $P(F)$. (F represents failure)}\\
\ \ \ \ F = F_1F_2F_3 \cup F_1F_2F_4 \cup\ ...\ \cup F_3F_4F_5\\
\ \ \ \ \Rightarrow P(F) \le P(F_1F_2F_3) + P(F_1F_2F_4) +\ ...\ + P(F_3F_4F_5) = \binom{5}{3} (0.001)^3 = \binom{5}{3} 10^{-9} = 10^{-8}\\
\ \ \ \ \textbf{Note}\ \text{# of failure is Binomial$(5,0.001)$, so we can compute $P(F)$ exactly}\\
\ \ \ \ \ \ \ \ P(F)=\binom{5}{3}p^3 (1-p)^2 + \binom{5}{4}p^4 (1-p) + \binom{5}{5}p^5\\
\ \\
\textbf{Problem 2.43}\ \text{(Failure of a cube) Each corner fails independently with probability $p=0.001$.}\\
\text{System fails if two neighboring nodes fail. Bound $P(F)$}\\
\ \ \ \ F = F_1F_2\cup F_1F_4\cup F_2F_3\cup\ ...\ \cup F_7F_8\\
\ \ \ \ P(F) \le \overbrace{P(F_1F_2)}^{p^2} + \ ...\ + \overbrace{P(F_7F_8)}^{p^2} = 12 \times (0.001)^2 = 12 \times 10^{-6}\ 
\text{(The number of edges is 12)}\\
$

\bullet \text{Each link $i$ has}\ \textbf{failure}\ \text{probability $p_i$}\ \text{and}\ \textbf{capacity}\ C_i\ (q_i = (1-p_i))\\

\ \ \ \ \text{F) What is the distribution of capacity?}\\
\ \ \ \ \ \ \ \ \text{Let $X$ be the network capacity}\\
\ \ \ \ \ \ \ \ \text{Capacity $X$ may have values $0,\ 10,\ 20,\ 30$}\\
\ \ \ \ \ \ \ \ P_X(0) = P(\text{outage}) = p_1p_2,\ P_X(10) = (1-p_1)p_2,\ P_X(20)=p_1(1-p_2),\ P_X(30) = (1-p_1)(1-p_2)\\
