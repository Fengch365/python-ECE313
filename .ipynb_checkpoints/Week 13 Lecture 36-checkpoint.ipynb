\textbf{Previously:}\ \text{Correlation and Covariance}\\
\bullet \mbox{correlation:}\ E[XY]\\
\bullet \mbox{covariance:}\ \mbox{Cov}(X,Y) = E[(X-E[X])(Y-E[Y])] = E[XY] - E[X]E[Y]\\
\bullet \mbox{correlation coefficient:}\ \rho_{XY} = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X) \text{Var}(Y)}}\ (|\rho_{XY}| \le 1)\\
\bullet \text{$X$ and $Y$ are uncorrelated if } \mbox{Cov}(X,Y) = 0,\ \text{equivalently,}\ E[XY] = E[X]E[Y]\\
\bullet \text{If $X$ and $Y$ are independent, then they are uncorrelated}\\
\ \ \ \text{Note: If $X$ and $Y$ are independent, } E[XY] = E[X]E[Y]\\
\bullet \text{linearity of covariance:}\\
\ \ \ \mbox{Cov}(X+Y,U+V) = \mbox{Cov}(X,U) + \mbox{Cov}(X,V) + \mbox{Cov}(Y,U) + \mbox{Cov}(Y,V)\\
\ \ \ \mbox{Cov}(aX+b,cY+d) = \mbox{Cov}(aX,cY) = a\cdot c\cdot \mbox{Cov}(X,Y)\\
\bullet \text{Schwarz Inequality: For RVs}\ X,\ Y,\ |E[XY]| \le \sqrt{E[X^2]E[Y^2]}\\
\ \ \ \text{If $E[X^2]\neq 0$, equality holds if and only if $Y=cX$ for some $c$}

\text{b) }E[X] = E[X_1+X_2+...+X_n] = \frac{n}{6}\\

\ \ \ \ \mbox{If}\ i=j,\ \mbox{Cov}(X_i, X_i) = \mbox{Var}(X_i) = \frac{5}{36}\\
\ \ \ \ \mbox{If}\ i \neq j,\ \mbox{Cov}(X_i, X_j) = 0\\
\ \ \ \ \text{($X_i$ and $X_j$ are independent $\implies$ they are uncorrelated)}\\
\ \ \ \ \mbox{Var}(X) = \sum_{i=1}^{n} \mbox{Cov}(X_i,X_i) = \frac{5n}{36}\\
\text{c) Cov}(X_i,Y_i)\\
\ \ \ \ \text{If $i\neq j$, two different dice rolls, $X_i$ and $Y_j$ are independent} \implies \mbox{Cov}(X_i,Y_i) = 0\\
\ \ \ \ \text{If $i=j$, $X_i$ and $Y_i$ are not independent}\\
\ \ \ \ \ \ \ \ \mbox{Cov}(X_i,Y_i) = E[X_iY_i] - \underbrace{E[X_i]}_{\frac{1}{6}}\underbrace{E[Y_i]}_{\frac{1}{6}}\\
\ \ \ \ \ \ \ \ \ \ \ \ X_iY_i = 1\ \mbox{if}\ X_i=Y_i=1\ \mbox{But this is impossible}\\
\ \ \ \ \ \ \ \ \implies X_iY_i = 0\implies E[X_iY_i]=0\implies \mbox{Cov}(X_i,Y_i) = -\frac{1}{36}\\
\text{d) Cov}(X,Y) = \mbox{Cov}(\sum_{i=1}^{n} X_i, \sum_{i=1}^{n} Y_i) = \sum_{i=1}^{n} \sum_{j=1}^{n} \mbox{Cov}(X_i, Y_j)
= \sum_{i=1}^{n} \mbox{Cov}(X_i, Y_i) = -\frac{n}{36}\\
\text{e) }\rho_{XY} = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)\text{Var}(Y)}}
= \frac{-\frac{n}{36}}{\sqrt{(\frac{5n}{36})(\frac{5n}{36})}} = -\frac{1}{5}\\
\ \ \ \ \implies X,\ Y\ \text{are negatively correlated}

$
\textbf{Minimum Mean Square Error (MMSE) Estimation}\\
\text{Suppose I want to estimate $Y$, but I observe $X$ instead. If $X$ and $Y$ are related}\\
\text{through a joint distribution (e.g. $F_{XY}$), I can try to estimate $Y$ from $X$.}\\
\ \\
\text{We will use the MMSE criterian.}\\
\text{I want a function $f(X)$ (my estimator) that minimizes } \underbrace{E[(Y-f(X))^2]}_{\text{mean square error}}\\
\ \\
\text{Quick detour: estimate $Y$ with a constant $\delta$}\\
\ \ \ \ \text{Intuitively, a good constant estimator for $Y$ is $E[Y]$.}\\
\ \ \ \ \text{Is that the MMSE estimator?}\\
\ \\
\ \ \ \ \text{We want $\delta$ that minimizes}\ E[(Y-\delta)^2] = E[Y^2] - 2\delta E[Y] + \delta^2\\
\ \ \ \ \text{Differentiate with respect to $\delta$}\\
\ \ \ \ \frac{d}{d\delta} E[(Y-\delta)^2] = -2E[Y] + 2\delta = 0\\
\ \ \ \ \implies \delta = E[Y] \implies E[Y]\ \text{is the constant MMSE estimator}\\
\ \ \ \ \text{Notice that MSE}\ = E[(Y-\delta)^2] = E[(Y-E[Y])^2] = \mbox{Var}(Y)
$
