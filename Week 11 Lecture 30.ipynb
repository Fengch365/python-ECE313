\textbf{Previously}\\
\bullet \text{Jointly continuous-type (they have joint pdf $f_{X,Y}(u,v)$)}\\
\bullet \text{marginalization}\ f_X(u) = \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dv\\
\bullet \text{conditional pdf}\ f_{Y|X}(v|u) = \frac{f_{X,Y}(u,v)}{f_X(u)}\ \text{(define only if $f_X{u}>0$)}\\
\bullet \text{Jointly uniform distributed RVs}\\
\ \ \ \text{For some region $S\subset \mathbb{R}^2$,}\\
\ \ \ \ \ \ \ f_{X,Y}(u,v) = \begin{cases}
\frac{1}{\text{area of $S$}} & \text{if $(u,v)\in S$}\\
0 & \text{otherwise}
\end{cases}\\
\ \ \ \text{For any region $A\subset \mathbb{R}^2$,}\\
\ \ \ \ \ \ \ P\{(X,Y)\in A\} = \frac{\text{area of $A \cap S$}}{\text{area of $S$}}
$

$
\text{area of $S$:}\ 1^2 - (\frac{1}{2})^2 = \frac{3}{4}\\
f_{X,Y}(u,v) = \begin{cases}
\frac{4}{3} & \text{if $(u,v)\in S$}\\
0 & \text{otherwise}
\end{cases}
$


$
\text{a)}\ f_X(u_0) = \int_{-\infty}^{\infty} f_{X,Y}(u_0,v)\ dv = \begin{cases}
0 & u_0<0\\
\int_{\frac{1}{2}}^{1} \frac{4}{3}\ dv = \frac{4v}{3}|_{\frac{1}{2}}^{1} = \frac{2}{3} & 0 \le u_0 < \frac{1}{2}\\
\int_{0}^{1} \frac{4}{3}\ dv = \frac{4v}{3}|_{0}^{1} = \frac{4}{3} & \frac{1}{2} < u_0 \le 1\\
0 & u_0>1
\end{cases}\\
\ \ \ \ \text{By symmetry, $f_Y$ is the same as $f_X$}
$


$
\text{b)}\ f_{Y|X}(v|u_0) = \frac{f_{X,Y}(u_0,v)}{f_X(u_0)}\ \text{(only defined $0<u_0<1$)}\\
\ \ \ \ \text{f_{X,Y}(u_0,v) is a "slice" of $f_{X,Y}$ at $u=u_0$} 
$

$
\ \ \ \ \Rightarrow f_{Y|X}(v|u_0) = \begin{cases}
\frac{4}{3}\cdot \frac{3}{2} = 2 & 0<u_0<\frac{1}{2},\ \frac{1}{2}<v<1\\
\frac{4}{3}\cdot \frac{3}{4} = 1 & \frac{1}{2}<u_0<1,\ 0<v<1\\
\end{cases}
$

$
\text{$X$ is uniformly distributed in}\ [0,1] \Rightarrow f_X(u) = \begin{cases}
1 & 0\le u \le 1\\
0 & \text{otherwise}
\end{cases}\\
\text{Given $X=u$, $Y$ is uniformly distributed in}\ [0,1-u]\\
f_{Y|X}(v|u) = \begin{cases}
\frac{1}{1-u} & \text{for}\ 0<v<1-u\\
0 & \text{otherwise}
\end{cases}\\
\text{find}\ f_{X,Y}(u,v):\\
\ \ \ \ f_{Y|X}(v|u) = \frac{f_{X,Y}(u,v)}{f_X(u)} \Rightarrow f_{X,Y}(u,v) = f_X(u)f_{Y|X}(v|u)\\
\ \\
\textbf{Note}:\ f_{X,Y}(u,v) = f_X(u)f_{Y|X}(v|u)\ \text{holds even when $f_X(u)=0$, because when $f_X(u)=0$, $f_{X,Y}(u,v)=0$ for any $v$}\\
\ \\
\Rightarrow f_{X,Y}(u,v) = \begin{cases}
\frac{1}{1-u} & 0<u<1,\ 0<v<1-u\\
0 & \text{otherwise}
\end{cases}\\
\text{For $0<v<1$,}\ f_Y(v) = \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dv = \int_{0}^{1-v} \frac{1}{1-u}\ du = -\ln (1-u)|_0^{1-v}
= -\ln (v) + \ln(1) = -\ln(v)
$

\textbf{Independence of RVs}\\
\text{Recall: Events $E$ and $F$ are independent if}\ P(EF) = P(E)P(F)\\
\ \\
\textbf{Definition}\ \text{RVs $X$ and $Y$ are independent if any pair of events of the form}\\
\text{$\{X\in A\}$ and $\{Y\in B\}$ (for any $A$, $B\subset \mathbb{R}$)},\\
P(\{X\in A\},\ \{Y\in B\}) = P\{X\in A\}P\{X\in B\}\\
\ \\
\textbf{Equivalent definitions:}\\
\bullet \text{$X$ and $Y$ are independent if and only if}\\
\ \ \ F_{X,Y}(u,v) = F_X(u)F_Y(v)\ \text{for all $u,v$}\\
\bullet \text{If $X$ and $Y$ are discrete-type, then they are independent}\\
\ \ \ \text{if and only if, for all $u,v$,}\ P_{XY}(u,v) = P_X(u)P_Y(v)\\
\bullet \text{If $X$ and $Y$ are jointly continuous-type, then they are independent}\\
\ \ \ \text{if and only if, for all $u,v$,}\ f_{XY}(u,v) = f_X(u)f_Y(v)\\
\bullet \text{If $X$ and $Y$ are jointly continuous-type, then they are independent}\\
\ \ \ \text{if and only if, for all $u$, either $f_X(u)=0$ or $f_{Y|X}(v|u)=f_Y(v)$ for all $v$}\\
\ \ \ \text{(In words, value of $X=u$ doesn't affect the conditional distribution of $Y$)}
