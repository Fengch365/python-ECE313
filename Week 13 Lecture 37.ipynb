$
\textbf{Previously:}\\
\ \ \ \text{Estimating a RV $Y$ from $X$}\\
\bullet \text{mean square error (MSE)}\\
\ \ \ \ \ \ \ \text{For an estimator $\hat{Y}$ of $Y$, MSE =}\ E[(\hat{Y}-Y)^2]\\
\bullet \text{minimum MSE (MMSE) estimator of $Y$ minimizes MSE}\\
\bullet \text{constant MMSE estimator of $Y$ is $E[Y]$}\\
\ \ \ \ \ \ \ \text{Among all constant $\delta$, $\delta = E[Y]$ minimizes $E[(Y-\delta)^2]$}\\
\ \ \ \ \ \ \ \text{Moreover, when $\delta = E[Y]$, the MSE is}\\
\ \ \ \ \ \ \ E[(Y-E[Y])^2] = \mbox{Var}(Y) \rightarrow \text{how difficult it is to estimate $Y$ with a constant}
$

$
\textbf{Minimum MSE estimation}\\
\text{Suppose I want to estimate $Y$, but I observe $X$. I want to find a function}\\
\text{$f(X)$ (my estimator of $Y$) that minimizes MSE =}\ E[(Y-f(X))^2]\\
\ \\
\textbf{unconstrained MMSE estimation:}\ \text{any function $f$ is allowed}\\
\text{Suppose I observe that $X=10$.}\\
\text{Since I know joint pdf $f_{XY}(u,v)$, I can find $f_{Y|X}(v|10)$}\\
\text{Natural guess for the estimator is:}\ E[Y|X=10] = =\int_{-\infty}^{\infty}\ v f_{Y|X}(v|10)\ dv\ \text{(conditional mean)}\\
\ \\
\text{As it turns out, the conditional mean $E[Y|X=10]$ is indeed the MMSE estimator of}\\
\text{$Y$ given $X=u$. We will write $g^*(u)=E[Y|X=u]$ and $\underbrace{g^*(X)}_{\text{RV}}=E[Y|X]$}\\
\ \\
\text{This means that amongs all function $g(X)$, $g^*(X)=E[Y|X]$ minimizes MSE}\\
\text{In other words,}\ E[(Y-E[Y|X])^2] \le E[(Y-g(X))^2]\ \text{for any $g$}\\
\ \\
\text{What is the corresponding MSE?}\\
E[(Y-E[Y|X])^2] = E[Y^2] - E[(E[Y|X])^2]\ \text{(see book)}
$

$
\textbf{Linear Estimators}\\
\text{In practice, computing $g^*(u)$ may be hard. Often, we restrict ourselves to linear estimators}\ L(X)=aX+b\\
\text{Linear MMSE estimator: we need $a,\ b$ that minimize}\\
E[(Y-L(X))^2] = E[(Y-aX-b)^2] = E[(\underbrace{(Y-aX)}_{\text{RV}}-b)^2]\\
\rightarrow \text{equivalent to constant MMSE estimator $Y-aX$. we know $b=E[Y-aX] = E[Y]-aE[X]$}\\
\implies L(X)=aX+E[Y]-aE[X] = a(X-E[X]) + E[Y]\\
E[(Y-L(X))^2] = E[((\underbrace{Y-E[Y]}_{\tilde{Y}})-a(\underbrace{X-E[X]}_{\tilde{X}}))^2]
\ \ \ \ = E[\tilde{Y}^2 - 2a \tilde{Y}\tilde{X} + a^2 \tilde{X}^2] = E[\tilde{Y}^2] - 2a E[\tilde{X}\tilde{Y}] + a^2E[\tilde{X}^2]\\
\ \ \ \ = \mbox{Var}(Y) - 2a \mbox{Cov}(X,Y) + a^2 \mbox{Var}(X)\\
\frac{d}{da} E[(Y-L(X))^2] = -2 \mbox{Cov}(X,Y) + 2a \mbox{Var}(X) = 0 \implies a = \frac{\text{Cov}(X,Y)}{\text{Var}(X)}\\
\implies L^*(X) = \frac{\text{Cov}(X,Y)}{\text{Var}(X)} (X-E[X]) + E[Y]\\
\text{MSE} = E[(Y-L^*(X))^2] = \mbox{Var}(Y) - \frac{2\text{Cov}(X,Y)}{\text{Var}(X)} \text{Cov}(X,Y) 
+ (\frac{\text{Cov}(X,Y)}{\text{Var}(X)})^2 \text{Var}(X)
= \text{Var}(Y) - \frac{\text{Cov}(X,Y)^2}{\text{Var}(X)}\\
\ \\
\textbf{Notes:}\\
\bullet \text{We will sometimes denote $L^*(X) = \hat{E}[Y|X]$ and call this "wide-sense conditional expectation"}\\
\bullet \text{An equivalent formula for $\text{Var}(Y) - \frac{\text{Cov}(X,Y)^2}{\text{Var}(X)}$ is MSE}
=E[Y^2] - E[\hat{E}[Y|X]^2]\\
\ \ \ \text{(compare with $E[Y^2]-E[E[Y|X]^2]$) for unconstrained MMSE}\\
\bullet \text{we saw three estimators for $Y$: $\underbrace{E[Y]}_{\text{worse}},\ \hat{E}[Y|X],\ \underbrace{E[Y|X]}_{\text{best}}$}\\
\ \ \ \text{their MSE satisfy}\ E[(Y-E[Y])^2] \ge E[(Y-\hat{E}[Y|X])^2] \ge E[(Y-E[Y|X])^2]
$

$
\text{a) } g^*(u) = E[Y|X=u] = \int_{-\infty}^{\infty}\ v f_{Y|X}(v|u)\ dv\\
\ \ \ \ f_{Y|X}(v|u) = \frac{f_{XY}(u,v)}{f_X(u)}\ \text{if}\ f_X(u) > 0\\
\ \ \ \ f_X(u) = \int_{-\infty}^{\infty}\ f_{XY}(u,v)\ dv = \begin{cases}
\int_{\frac{u+1}{2}}^{u+1}\ 2\ dv = 2\cdot \frac{u+1}{2} = u+1 & -1<u<0\\
\int_{\frac{u+1}{2}}^{1}\ 2\ dv = 2\cdot (1-\frac{u+1}{2}) = 1-u & 0<u<1
\end{cases}
$


$
\ \ \ \ f_{Y|X}(v|u) = \begin{cases}
\frac{2}{1+u} & -1<u<0,\ \frac{u+1}{2} < v < u+1\\
\frac{2}{1-u} & 0 < u < 1,\ \frac{u+1}{2} < v < 1
\end{cases}\\
\ \ \ \ E[Y|X=u] = \begin{cases}
\int_{-\infty}^{\infty}\ v f_{Y|X}(u|v)\ dv = \int_{\frac{u+1}{2}}^{u+1}\ v\cdot \frac{2}{1+u}\ dv = \frac{3(1+u)}{4} & -1<u<0\\
\int_{\frac{u+1}{2}}^{1}\ v\cdot \frac{2}{1-u}\ dv = \frac{3+u}{4} & 0 < u < 1
\end{cases}
$


$
\ \ \ \ \text{To find MSE: MSE}= E[Y^2] - E[E[Y|X]^2],\ \mbox{where}\ E[Y^2] = \int_{-\infty}^{\infty}\ v^2 f_Y(v)\ dv
$

$
\ \ \ \ \text{$f_Y$ can be found via $\int_{-\infty}^{\infty} f_{XY}(u,v)\ du$}
$


$
\ \ \ \ E[{\underbrace{E[Y|X]}_{\text{function of $X$}}}^2] = \int_{-1}^{1}\ E[Y|X=u]^2 f_X(u)\ du\\
\ \ \ \ \mbox{MSE} = E[Y^2] - E[E[Y|X]^2] = \frac{1}{96}\\
\text{b) Linear MMSE } \hat{E}[Y|X] = \frac{\text{Cov}(X,Y)}{\text{Var}(X)} \cdot (X-\underbrace{E[X]}_{=0}) + E[Y]\\
\ \ \ \ E[Y] = \int_{-\infty}^{\infty}\ v f_Y(v)\ dv = \int_{0}^{1}\ v (2v)\ dv = \frac{2v^3}{3}|_0^1 = \frac{2}{3}\\
\ \ \ \ \mbox{Cov}(X,Y) = E[XY] - \underbrace{E[X]}_{=0}E[Y] = E[XY] = \int_{v=0}^{1}\int_{u=v-1}^{2v-1}\ 2(uv)\ dudv = \frac{1}{2}\\
\ \ \ \ \mbox{Var}(X) = E[X^2] - \underbrace{E[X]^2}_{=0} = \int_{-1}^{1}\ u^2 f_X(u)\ du = \frac{1}{6}\\
\ \ \ \ \implies \hat{E}[Y|X] = \frac{1/12}{1/6}\cdot X + \frac{2}{3} = \frac{1}{2}X + \frac{2}{3}
$
