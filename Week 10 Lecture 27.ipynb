\text{Binary hypothesis testing for continuous-type}\\
\ \ \ \ \text{Two hypotheses}\\
\ \ \ \ \ \ \ \ \text{$H_0$: $X$ has pdf $f_0$}\\
\ \ \ \ \ \ \ \ \text{$H_1$: $X$ has pdf $f_1$}\\
\ \ \ \ \text{We observe $X=u$ and pick $H_0$ or $H_1$}\\
\ \ \ \ \text{Likelihood ratio}\ \Lambda(u) = \frac{f_1(u)}{f_0(u)}\\
\ \ \ \ \text{Decision rules}\\
\ \ \ \ \ \ \ \ \text{ML:} \Lambda(u)\ \begin{cases}
> 1 & \text{pick $H_1$}\\
< 1 & \text{pick $H_0$}\\
\end{cases}\\
\ \ \ \ \ \ \ \ \text{MAP:} \Lambda(u)\ \begin{cases}
> \frac{\pi_0}{\pi_1} & \text{pick $H_1$}\\
< \frac{\pi_0}{\pi_1} & \text{pick $H_0$}\\
\end{cases}\\
$

$
\text{pdfs:}\ H_0:\ X \text{ is}\ N(0,a^2)\Rightarrow f_0(u)=\frac{1}{\sqrt{2\pi a^2}} e^{-\frac{u^2}{2a^2}}\\
\ \ \ \ \ \ \ \ \ H_1:\ X \text{ is}\ N(0,b^2)\Rightarrow f_1(u)=\frac{1}{\sqrt{2\pi b^2}} e^{-\frac{u^2}{2b^2}}\\
\textbf{ML decision rule}\\
\Lambda(u)=\frac{f_1(u)}{f_0(u)}=\frac{\sqrt{2\pi a^2}}{\sqrt{2\pi b^2}} \frac{e^{-\frac{u^2}{2b^2}}}{e^{-\frac{u^2}{2a^2}}}
= \frac{a}{b} exp(-\frac{u^2}{2b^2}+\frac{u^2}{2a^2})\ \gtrless 1\\
\Leftrightarrow exp(\frac{u^2}{2}(\frac{1}{a^2}-\frac{1}{b^2})) \gtrless \frac{b}{a}\\
\Leftrightarrow \frac{u^2}{2}(\frac{b^2-a^2}{a^2 b^2})\gtrless \ln \frac{b}{a}\\
\Leftrightarrow u^2 \gtrless \underbrace{\frac{2 a^2 b^2}{b^2-a^2}}_{>0} \underbrace{\ln \frac{b}{a}}_{>0}\\
\Leftrightarrow |u| \gtrless \underbrace{\sqrt{\frac{2 a^2 b^2}{b^2-a^2}\ln \frac{b}{a}}}_{\gamma_{ML}}\\
|X| \begin{cases}
> \gamma_{ML} & \text{choose $H_1$}\\
< \gamma_{ML} & \text{choose $H_0$}\\
\end{cases}\\
$

\textbf{Jointly Distribution Random Variables}\\
\text{If $X$ and $Y$ are not (necessarily) independent, how do we decide their joint distribution?}\\
\ \\
\textbf{Joint CDF}\\
\text{Suppose $X$ and $Y$ are defined on the same probability space $(\Omega,F,P)$}\\
\text{(this is important for them to have a joint distribution)}\\
\text{Joint CDF of $X$ and $Y$ is defined as}\ F_{X,Y}(u,v) = P(X\le u, Y\le v)
$


$
\text{Can use joint CDF to express probability for any rectangul region?}\\
P((X,Y)\in R) = F_{X,Y}(u_2,v_2) - F_{X,Y}(u_1,v_2) - F_{X,Y}(u_2,v_1) + F_{X,Y}(u_1,v_1)
$


$
\textbf{Note}\\
\bullet \text{The probability of a non-rectangular shape is also determined by joint CDF using limits}\\
\bullet \text{Individual CDFs are also determined by joint CDF}\\
\ \ \ F_X(u) = P(X\le u) = P(X\le u, Y\le \infty) = \underbrace{F_{X,Y}(u,\infty)}_{\text{means}\ \lim_{v\to \infty} F_{X,Y}(u,v)}\\
\ \ \ \text{Similarly,}\ F_Y(v) = F_{X,Y}(\infty,v)
$

$
\textbf{Properties of joint CDF}\ (F=F_{X,Y})\\
\bullet 0\le F(u,v) \le 1\\
\bullet F(u,v)\ \text{is non-decreasing in $u$ (and in $v$)}\\
\bullet F(u,v)\ \text{is right continuous in $u$ (and in $v$)}\\
\bullet \text{If $a<b$, $c<d$,}\ F(b,d) - F(b,c) - F(a,d) + F(a,c) \ge 0\\
\bullet \lim_{u\to -\infty} F(u,v)=0\ \text{for all $v$,}\ \lim_{v\to -\infty} F(u,v)=0\ \text{for all $u$}\\
\bullet \lim_{u\to \infty}\lim_{v\to \infty} F_{X,Y}(u,v) = 1\\
\ \\
\textbf{Ex}\ \text{Suppose $X$ and $Z$ are independent Bernoulli$(\frac{1}{2})$ RVs.}\\
\text{Let $Y=X+Z$. Find $F_{X,Y}(u,v)$}\\
\ \ \ \ \text{Note: $(X,Y)$ can take values}\ (0,0),\ (0,1),\ (1,1),\ (1,2).\ \text{Each has probability}\ \frac{1}{4}
$


$
\textbf{Joint probability mass function (pmf)}\\
\text{If $X$ and $Y$ are discrete-type RVs (on the same probability space), they have a joint pmf:}\ p_{X,Y}(u,v) = P(X=u,Y=v)\\
\ \\
\text{The}\ \textbf{marginal}\ \text{pmfs $p_X$ and $p_Y$ can be obtained from $p_{X,Y}$:}\\
p_X(u) = P(X=u) = P(X=u, Y\in \underbrace{\{v_1,v_2,...\}}_{\text{all possible values of $Y$}})\\
\ \ \ \ \ \ \ \ \ \ = \sum_j P(X=u,Y=v_j) = \sum_j p_{X,Y} (u,v_j)\\
p_Y(v) = \sum_i p_{X,Y}(u_i,v)\\
\ \\
\text{We also define the}\ \textbf{conditional}\ \text{pmf:}\\
P_{Y|X}(u|v) = P(Y=u|X=v) = \frac{P(Y=u,X=v)}{P(X=v)} = \frac{p_{X,Y}(v,u)}{p_X(v)} = \frac{p_{X,Y}(v,u)}{\sum_i p_{X,Y}(v,u_i)}
$
