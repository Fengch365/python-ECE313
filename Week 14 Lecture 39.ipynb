\textbf{Previously:}\\
\bullet \text{Law of large numbers (LLN):}\\
\ \ \ \text{If $X_1,\ X_2,\ ...$ are uncorrelated RVs with $E[X_i]=\mu$ for $i=1,2,...$}\\
\ \ \ \text{and $\mbox{Var}(X_i)\le c$ for $i=1,2,...$, then, for all $\delta > 0$,}\\
\ \ \ P\{|\frac{S_n}{n}-\mu| > \delta\} \le \frac{c}{n\delta^2}\to 0\ \mbox{as}\ n\to \infty\ \mbox{where}\ S_n = \sum_{i=1}^{n}X_i\\
\ \ \ \text{Intuitively, $\frac{S_n}{n} \approx \mu$ for large $n$}\\
\bullet \text{Central Limit Theorem:}\\
\ \ \ \text{If $X_1,\ X_2,\ ...$ are iid RVs with $E[X_1]=\mu$ and $\mbox{Var}(X_1)=\sigma^2 < \infty$,}\\
\ \ \ \text{then, for all $c$,}\ \lim_{n\to \infty} P\{\frac{S_n-n\mu}{\sqrt{n\sigma^2}}\le c\} 
= \underbrace{\Phi(c)}_{\text{CDF of $N(0,1)$}}\\
\ \ \ \text{Intuitively, $\frac{S_n-n\mu}{\sqrt{n\sigma^2}} \approx N(0,1)$ for large $n$}\\
\ \ \ \text{For special case where $E[X_i]=0$ and $\mbox{Var}(X_i)=1$,}\\
\ \ \ \ \ \ \ \text{LLN: $\frac{S_n}{n} \approx 0$}\\
\ \ \ \ \ \ \ \text{CLT: $\frac{S_n}{\sqrt{n}} \approx N(0,1)$}

\textbf{Joint Gaussian Distribution}\\
\text{Suppose $(U_i,V_i)$ is the height and weight of the $i^{th}$ student, $i=1,2,3,...,n$.}\\
\ \ \ \ \frac{1}{\sqrt{n}} \sum_{i=1}^{n}(U_i,V_i) = (\frac{\sum U_i}{\sqrt{n}}, \frac{\sum V_i}{\sqrt{n}})\\
\text{By CLT, both $\frac{\sum U_i}{\sqrt{n}}$ and $\frac{\sum V_i}{\sqrt{n}}$ are approximately Gaussian.}\\
\text{Moreover, for constants $a,b$, I define $Z_i=aU_i+bV_i$ and $\frac{Z_i}{\sqrt{n}} = \frac{\sum aU_i+bV_i}{\sqrt{n}}$,}\\
\text{by CLT, is also approximately Gaussian.}\\
\text{This motivates the notion of jointly Gaussian RVs.}\\
\textbf{Definition:}\ \text{RVs $X$ and $Y$ are jointly Gaussian if $aX+bY$ is Gaussian for all $a,b \in \mathbb{R}$}\\
\ \ \ \ \text{Note:}\\
\ \ \ \ \bullet \text{By picking $a=1,b=0\ (\mbox{or}\ a=0,b=1)$, we that $X$ and $Y$ must be Gaussian}\\
\ \ \ \ \bullet \text{A constant $X=c$ can be seen as a $N(c,0)$ RV}\\
\ \ \ \ \bullet \text{$X$ and $Y$ have the bivariate normal distribution}\\
\text{If $X,Y$ are jointly Gaussian, do they have a joint pdf?}\\
\ \ \ \ \text{Not always. If $X$ is $N(\mu_X, {\sigma_X}^2)$ and $Y=cX+d$, then the support}\\
\ \ \ \ \text{of $(X,Y)$ is just a line. So there cannot be a joint pdf ("no volume")}\\
\ \ \ \ \text{Notice that in this case,}\ \rho_{XY} = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y} = 1\ \mbox{or}\ -1\ 
\text{(depends on whether $c>0$)}
$


$
\text{But in all other cases (i.e., $\rho_{XY} \neq 1$ and $\rho_{XY}\neq -1$), $X$ and $Y$ have a joint pdf,}\\
\text{given by}\ f_{XY}(u,v) = \frac{1}{2\pi \sigma_X \sigma_Y \sqrt{1-\rho^2}} 
exp(-\frac{(\frac{u-\mu_X}{\sigma_X})^2 + (\frac{v-\mu_Y}{\sigma_Y})^2 - 2\rho (\frac{u-\mu_X}{\sigma_X})(\frac{v-\mu_Y}{\sigma_Y})}{2(1-\rho^2)}),\\
\text{where}\ \mu_X = E[X],\ \mu_Y = E[Y],\ \sigma_X = \sqrt{\text{Var}(X)},\ \sigma_Y = \sqrt{\text{Var}(Y)}\ \mbox{and}\ \rho 
= \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}\\
\ \\
\text{special case: standard bivariate normal distribution: $X,Y$ are each $N(0,1)$ and independent}\\
f_{XY}(u,v) = f_X(u)f_Y(v) = \frac{1}{\sqrt{2\pi}} e^{-\frac{u^2}{2}}\cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{v^2}{2}}
= \frac{1}{2\pi} e^{-\frac{u^2+v^2}{2}},\\
\text{which is a special case of $f_{XY}$ above for}\ \mu_X=\mu_Y=0,\ \sigma_X=\sigma_Y=1,\ \mu=0
