$
\textbf{Previously}\ \text{joint pdf}\\
\bullet \text{If $X,Y$ are jointly continuous-type, they have a joint pdf $f_{X,Y}(u,v)$ and $F_{X,Y}(u_0,v_0)$ can}\\
\ \ \ \text{be computed as}\ F_{X,Y}(u_0,v_0) = \int_{-\infty}^{u_0} \int_{-\infty}^{v_0} f_{X,Y}(u,v)\ dudv\\
\bullet \text{For any region $R \subseteq \mathbb{R}^2$,}\ P\{(X,Y)\in \mathbb{R}\} = \iint_R f_{X,Y}(u,v)\ dvdu\\
$

$
\bullet\ \text{marginalization:}\ f_X(u)=\int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dv\\
\bullet \text{conditional pdf:}\ f_{Y|X}(v|u_0) = \underbrace{\frac{f_{X,Y}(u_0,v)}{f_X(u_0)}}_{\text{pdf in $v$}}\ 
\text{(defined if $f_X(u_0)>0$)}
$

$
\text{a) $f_{X,Y}$ must integrate to 1}\\
\ \ \ \ 1 = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dudv = \int_{u=0}^{1} \int_{v=0}^{1-u} c(1-u-v)\ dvdu\\
\ \ \ \ = c \int_0^1 (v-uv-\frac{v^2}{2})_0^{1-u}\ du = c \int_0^1 (\overbrace{1-u-u+u^2}^{(1-u)^2} - \frac{(1-u)^2}{2})\ du\\
\ \ \ \ = c \int_0^1 \frac{(1-u)^2}{2}\ du = \frac{c}{2} \int_0^1 (u^2-2u+1)\ du\\
\ \ \ \ = \frac{c}{2} (\frac{u^3}{3}-u^2+u)_0^1 = \frac{c}{6}\Rightarrow c=6\\
\text{b)}\ f_X(u) = \int_{-\infty}^{\infty} f_{X,Y}(u,v)\ dv = 0\ \text{if $u<0$ or $u>1$}\\
\ \ \ \ \text{If $0<u<1$,}\ f_X(u) = \int_0^{1-u} 6(1-u-v)\ dv = 6(v-uv-\frac{v^2}{2})_0^{1-u} = 
\ \ \ \ 6(\underbrace{1-u-u+u^2}_{(1-u)^2}-\frac{(1-u)^2}{2}) = 6\cdot \frac{(1-u)^2}{2} = 3\cdot (1-u)^2\\
\text{c)}\ f_{Y|X}(v|u_0) = \frac{f_{X,Y}(u_0,v)}{f_X(u_0)}\ \text{(only defined if $0<u_0<1$)}\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = 0 \ \text{if $v>1-u_0$ or $v<0$}\\
\ \ \ \ \text{If $0<v<1-u_0$,}\ f_{Y|X}(v|u_0) = \frac{6(1-u_0-v)}{3(1-u_0)^2} = 2\cdot \frac{1-u_0-v}{(1-u_0)^2}
$

\textbf{Uniform joint pdfs}\\
\text{Let $S\subseteq \mathbb{R}^2$ (region on x-y plane) with finite area.}\\
\text{The uniform distribution over $S$ has joint pdf}\\
\ \ \ \ f_{X,Y}(u,v) = \begin{cases}
\frac{1}{\text{area of $S$}} & \text{for $(u,v)\in S$}\\
0 & \text{for $(u,v)\notin S$}
\end{cases}\\
\text{The probability of the event $\{(x,y) \in A\}$ is given by}\\
P\{(X,Y)\in A\} = \iint_A f_{X,Y}(u,v)\ dudv = \iint_{A \cap S} \frac{1}{\text{area of $S$}}\ dvdu =
\frac{1}{\text{area of $S$}} \iint_{A \cap S} \ dvdu = \frac{\text{area of $A \cap S$}}{\text{area of $S$}}
$

$
\text{a)}\ A=\{(u,v):\ u\ge 0,\ v\ge 0\}\\
\ \ \ \ P((X,Y)\in A) = \frac{\text{area of $A \cap S$}}{\text{area of $S$}} = \frac{1}{4}
$


$
\text{b)}\ P\{X^2+Y^2\le r^2\} = P\{(X,Y)\in \underbrace{\{(u,v):\ u^2+v^2\le r^2\}}_{\text{$B$ circle with radius $r$}}\}\\
\ \ \ \ \text{If}\ r<1:\ P\{(X,Y)\in B\} = \frac{\text{area of $B \cap S$}}{\text{area of $S$}} 
= \frac{\text{area of $B$}}{\text{area of $S$}} = \frac{\pi r^2}{\pi\cdot 1^2} = r^2\\
\ \ \ \ \text{If}\ r\ge 1:\ P\{(X,Y)\in B\} = \frac{\text{area of $\overbrace{B \cap S}^{S}$}}{\text{area of $S$}} = 1
$


$
\text{c)}\ f_X(u) = \int_{-\infty}^{\infty} f_{X,Y} (u,v)\ dv = 0\ \text{if $|u|>1$}\\
\ \ \ \ \text{If}\ -1<u<1:\ f_X(u) = \int_{-\sqrt{1-u^2}}^{\sqrt{1-u^2}} \frac{1}{\pi}\ dv
= \frac{v}{\pi}|_{-\sqrt{1-u^2}}^{\sqrt{1-u^2}} = \frac{2\sqrt{1-u^2}}{\pi}
$


$
\text{d)}\ f_{Y|X}(v,u_0) = \frac{f_{X,Y}(u_0,v)}{f_X(u_0)}\ \text{(only defined for $-1<u_0<1$)}=
\begin{cases}
\frac{1}{\pi}\cdot \frac{\pi}{2\sqrt{1-u_0^2}} = \frac{1}{2\sqrt{1-u_0^2}} & |v| < \sqrt{1-u_0^2}\\
0 & |v| > \sqrt{1-u_0^2}
\end{cases}
$


$
\text{a)}\ f_X(u_0) = \begin{cases}
\int_{-\infty}^{\infty} f_{X,Y}(u_0,v)\ dv = \int_0^{u_0} e^{-u_0}\ dv = e^{-u_0}\cdot v|_{0}^{u_0} = u_0 e^{-u_0} & u_0>0\\
0 & u_0\le 0
\end{cases}\\
\ \ \ \ f_Y(v_0) = \begin{cases}
\int_{-\infty}^{\infty} f_{X,Y}(u,v_0)\ du = \int_{v_0}^{\infty} e^{-u}\ du = -e^{-u}|_{v_0}^{\infty} = e^{-v_0} & v_0>0\\
0 & v_0\le 0
\end{cases}\\
\text{b)}\ f_{Y|X}(v|u_0) = \frac{f_{X,Y}(u_0,v)}{f_X(u_0)}\ \text{(defined $u_0>0$)} =
\begin{cases}
\frac{e^{-u_0}}{u_0 e^{-u_0}} = \frac{1}{u_0} & 0 < v < u_0\\
0 & \text{otherwise}
\end{cases}\\
\ \ \ \ f_{X|Y}(u|v_0) = \frac{f_{X,Y}(u,v_0)}{f_X(v_0)}\ \text{(defined $v_0>0$)} =
\begin{cases}
\frac{e^{-u}}{e^{-v_0}} = e^{-(u-v_0)} & u > v_0\\
0 & u \le v_0
\end{cases}\\
$
