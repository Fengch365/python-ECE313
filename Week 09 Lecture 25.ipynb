\ \ \ \text{$X$ has a known distribution. $Y=g(X)$. Find distribution of $Y$}\\
$


$
\text{1) $X$ is uniformly distributed in $[0,3]$} 
$

$
\text{CDF of $X$}
$


$
Y = (X-1)^2.\ \text{Support of $Y$?}\\
Y \in [0,4]
$


$
\text{2) CDF of $Y$? If $c\le 0$, $F_Y(c)=0$. If $c\ge 4$, $F_Y(c)=1$}\\
\ \ \ \ \text{If $c\in [0,1]$:}\\
\ \ \ \ \ \ \ \ F_Y(c) = P\{Y\le c\} = P\{1-\sqrt{c}\le X\le 1+\sqrt{c}\} = F_X(1+\sqrt{c})-F_X(1-\sqrt{c})\\
\ \ \ \ \ \ \ \ = \frac{1+\sqrt{c}}{3} - \frac{1-\sqrt{c}}{3} = \frac{2\sqrt{c}}{3}\\
\ \ \ \ \text{If $c\in [1,4]$:}\\
\ \ \ \ \ \ \ \ F_Y(c) = P\{Y\le c\} = P\{1-\sqrt{c}\le X\le 1+\sqrt{c}\} = F_X(1+\sqrt{c})-F_X(1-\sqrt{c}) = \frac{1+\sqrt{c}}{3} - 0
= \frac{1+\sqrt{c}}{3}
$


$
\text{1) Since}\ Y=F_X(X)\in [0,1]
$


$
\text{2) CDF of $Y$. For}\ c < 0, F_Y(c) = 0.\ \text{For}\ c > 1, F_Y(c) = 1.\\
\ \ \ \ \text{For}\ 0\le c\le 1,\ F_Y(c) = P\{Y\le c\} = P\{F_X(X)\le c\} 
= \underbrace{P\{F_X^{-1}(F_X(X))\le F_X^{-1}(c)\}}_{\text{CDF is non-decreasing and invertible}}
= \underbrace{P\{X\le F_X^{-1}(c)\}}_{\text{CDF of $X$ at $F_X^{-1}(c)$}} = 
F_X(F_X^{-1}(c)) = c
$


$
\textbf{Note:}\ \text{Above, we assumed that $F_X$ is invertible. In general, $F_X$ can have "flat" parts}\\
\ \ \ \ \ \ \ \ \ \ \text{In this case, we define}\ F_X^{-1}(u) = \text{min}\{c:\ F_X(c)\ge u\}\\
\ \ \ \ \ \ \ \ \ \ \text{Using this, we can show that}\ \textbf{$Y=F_X(X)$ is uniformly distributed in $(0,1)$ for any continuous-type RV $X$.}
$


$
\textbf{Generating a RV with a specified distribution}\\
\ \\
\text{I want to generate $X$ with a given CDF $F_X$.}\\
\text{From previous example, we know that $Y=F_X(X)$ is uniformly distributed in $(0,1)$.}\\
\text{Hence,}\ F_X^{-1}(Y) = F_X^{-1}(F_X(X)) = X \Rightarrow X = F_X^{-1}(\underbrace{Y}_{\text{uniformly distributed in $(0,1)$}})\\
\text{If I have $Y$ uniformly distributed, in $(0,1)$ (often available in a computer), I can generate $X=F_X^{-1}(Y)$ for any CDF $F_X$}
$


$
\text{We want $X=g(U)$ to have CDF}\ F_X(t) = \begin{cases}
1-e^{-t} & t\ge 0\\
0 & t < 0\\
\end{cases} \Rightarrow \text{Want to find $F_X^{-1}(U)$}\\
\text{For}\ 0 < U < 1,\ 1-e^{-t} = U \Rightarrow e^{-t} = 1 - U \Rightarrow t = - \ln (1-U)\\
\Rightarrow F_X^{-1}(U) = -\ln (1-U),\ 0 < U < 1\\
\textbf{Conclusion:}\ \text{Set}\ g(U) = - \ln (1-U).\\ 
\text{If $U$ is uniformly distributed in $(0,1)$, then $X=g(U)$ is Exponential$(1)$.}

\textbf{Failure Rate Functions}\\
\ \\
\text{Idea: failure rate for an electronic component}
$

$
\text{Suppose time until failure is a RV $T$}\\
\textbf{Definition:}\ \text{Failure Rate function is given by}\ h(t) = \lim_{\varepsilon \to 0} 
\frac{P(t<T \le t+\varepsilon\ | \ T > t)}{\varepsilon}\\
\text{If $T$ has a CDF $F_T$ and pdf $f_T$. Then:}\\
h(t) = \lim_{\varepsilon \to 0} \frac{P(t<T \le t+\varepsilon\ ,\ T > t)}{P(T>t)\cdot \varepsilon}
= \lim_{\varepsilon \to 0} \frac{F_T(t+\varepsilon) - F_T(t)}{(1-F_T(t)) \cdot \varepsilon} = \frac{f_T(t)}{1-F_T(t)}
$
