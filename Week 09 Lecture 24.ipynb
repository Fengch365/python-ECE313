\bullet \text{Central Limit Theorem}\\
\ \ \ \text{Idea: Independent RVs $X_1, X_2, X_3, ...$ If they are not "different",}\ \sum_{i=1}^n X_i\ \text{is approximately Gaussian.}\\
\ \ \ \text{Focused on the case where $X_1, X_2, ...$ are Bernoulli$(p)$}\\
\bullet \text{DeMoivre-Laplace Limit Theorem: Let}\ S_{n,p} = \sum_{i=1}^n X_i.\ \text{($S_{n,p}$ is binomial distributed)}\\
\ \ \ \text{For any $c$,}\ \lim_{n\to \infty} P\{\underbrace{\frac{S_{n,p}-np}{\sqrt{np(1-p)}}}_{\text{standardized sum (mean 0, variance 1)}} \le c\} = \underbrace{\Phi(c)}_{\text{CDF of $N(0,1)$}}

\textbf{ML Parameter Estimation for Continuous-type RVs}\\
\ \\
\text{Recall: Example for discrete-type RVs. Let $X$ have a Geometric$(p)$ distribution, for unknown $p$.}\\
\text{You observe $X=k$. Find $\hat{p}_{ML}$.}\\
\ \ \ \ \text{pmf:}\ p_X(k) = (1-p)^{k-1}p \Rightarrow \text{$\hat{p}_{ML}$ obtained by finding $p$ that maximizes $p_X(k)$}\\
\ \\
\text{For continuous-type RVs, we maximize the pdf instead.}\\
$

$
\text{pdf of T:}\ f_T(t) = \begin{cases}
\lambda e^{-\lambda t} & t\ge 0\\
0 & t < 0\\
\end{cases}\\
\text{Find $\lambda$ that maximizes $f_T(t)$}\\
\frac{d}{d\lambda} (\lambda e^{-\lambda t}) = e^{-\lambda t} + \lambda (-t) e^{-\lambda t} 
= e^{-\lambda t} \underbrace{(1-\lambda t)}_{=0} = 0 \Rightarrow \hat{\lambda}_{ML} = \frac{1}{t}\\
\textbf{Note:}\ \text{This could have been "guessed" by setting mean $\frac{1}{\lambda}$ equal to the observation t.}\\
\text{However, this approach does not always work.}

\textbf{Function of a Random Variable}\\
\ \\
\text{Let $X$ be a continuous-type RV and $Y=g(X)$}\\
\text{What is the distribution of $Y$?} \Rightarrow \text{calculate CDF, pdf(if $Y$ is continuous-type), pmf(if $Y$ is discrete-type)}\\
\ \\
\textbf{General Approach:}\\
\text{1) Find support of $Y$. Determine if continuous-type or discrete-type}\\
\text{If $Y$ is discrete-type:}\\
\text{2) Find pmf of $Y$.}\\
\text{If $Y$ is continuous-type:}\\
\text{2) Find CDF of $Y$.}\\
\text{3) Differentiate CDF to find pdf}
