$
\textbf{Correlation and Covariance}\\
\text{Simpler ways to describe joint distribution of $X$ and $Y$}\\
\bullet \text{correlation:}\ E[XY]\\
\bullet \text{covariance:}\ \mbox{Cov}(X,Y) = E[(X-E[X])(Y-E[Y])]\\
\bullet \text{correlation coefficient:}\ \rho_{XY} = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X) \text{Var}(Y)}}
= \frac{\text{Cov}(X,Y)}{\sigma_X\cdot \sigma_Y}
$

$
\textbf{Notes}\\
\bullet \mbox{Cov}(X,X) = E[(X-E[X])^2] = \mbox{Var}(X)\\
\bullet \mbox{Cov}(X,Y) = E[(X-E[X])(Y-E[Y])] = E[XY - XE[Y] - YE[X] + E[X]E[Y]]\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = E[XY] - E[X]E[Y] - E[X]E[Y] + E[X]E[Y]\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = E[XY] - E[X]E[Y]\ \mbox{(alternative formula)}
$

$
\textbf{uncorrelatedness}\\
\bullet \text{$X$ and $Y$ are said to be uncorrelated if Cov}(X,Y) = 0\ \mbox{(equivalently, $\rho_{XY}=0$)}\\
\bullet \text{$X,Y$ are positively correlated if}\ \mbox{Cov}(X,Y) > 0 \ (\rho_{XY}>0)\\
\bullet \text{$X,Y$ are negatively correlated if}\ \mbox{Cov}(X,Y) < 0 \ (\rho_{XY}<0)\\
\bullet \text{independence}\ \begin{matrix}
\Rightarrow \\
\nLeftarrow
\end{matrix}\
\text{uncorrelatedness}
$

$
\textbf{linearity of covariance}\\
\bullet \mbox{Cov}(X+Y, U+V) = \mbox{Cov}(X,U) + \mbox{Cov}(X,V) + \mbox{Cov}(Y,U) + \mbox{Cov}(Y,V)\\
\bullet \mbox{Cov}(aX+b, cY+d) = \mbox{Cov}(aX,cY) = a\cdot c\cdot \mbox{Cov}(X,Y)\\
\ \\
\text{This implies properties for variance:}\\
\bullet \mbox{Var}(aX+b) = \mbox{Cov}(aX+b,aX+b) = a^2 \mbox{Cov}(X,X) = a^2 \mbox{Var(X)}\\
\ \\
\text{If $X$ and $Y$ are uncorrelated $(\mbox{Cov}(X,Y)=0)$, then}\\
\bullet \mbox{Var}(X+Y) = \mbox{Cov}(X+Y, X+Y) = \mbox{Cov}(X,X) + \overbrace{\mbox{Cov}(X,Y)}^{=0} + \overbrace{\mbox{Cov}(Y,X)}^{=0}
+ \mbox{Cov}(Y,Y) = \mbox{Var}(X) + \mbox{Var}(Y)\\
\ \\
\text{If $X_1,X_2,...,X_n$ are pairwise uncorrelated $(\mbox{Cov}(X_i,X_j)=0\ \mbox{for}\ i\neq j)$,}\\
\bullet \mbox{Var}(\sum_{i=1}^{n} X_i) = \mbox{Cov}(\sum_{i=1}^{n} X_i, \sum_{i=1}^{n} X_i) 
= \sum_{i=1}^{n} \sum_{j=1}^{n} \mbox{Cov}(X_i, X_j) = \sum_{i=1}^{n} \mbox{Cov}(X_i, X_i) = \sum_{i=1}^{n} \mbox{Var}(X_i)
$

$
\mbox{a) Cov}(8X+3, 5Y-2) = \mbox{Cov}(8X,5Y) = 40 \mbox{Cov}(X,Y)\\
\mbox{b) Cov}(10X-5, -3X+15) = \mbox{Cov}(10X,-3X) = -30\mbox{Cov}(X,X) = -30\mbox{Var}(X)\\
\mbox{c) Cov}(X+2, 10X-3Y) = 10\mbox{Cov}(X,X) - 3\mbox{Cov}(X,Y) = 10\mbox{Var}(X)-3\mbox{Cov}(X,Y)\\
\mbox{d)}\ \rho_{10X,Y+4} = \frac{\text{Cov}(10X,Y+4)}{\sqrt{\text{Var}(10X)\text{Var}(Y+4)}}
= \frac{10 \text{Cov}(X,Y)}{\sqrt{10^2 \text{Var}(X)\text{Var}(Y)}}
= \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)\text{Var}(Y)}}
= \rho_{XY}
$

$
\textbf{Schwarz Inequality}\\
\text{For RVs, $X$ and $Y$,}\ |E[XY]| \le \sqrt{E[X^2]E[Y^2]}\\
\text{If $E[X^2]\neq 0$, equality holds ($|E[XY]| = \sqrt{E[X^2]E[Y^2]}$) if and only if $Y=cX$ for some constant $c$.}\\
\bullet \text{Consider RVs $\overbrace{X-E[X]}^{\tilde{X}}$ and $\overbrace{Y-E[Y]}^{\tilde{Y}}$. Then}\\
\ \ \ |E[(X-E[X])(Y-E[Y])]| \le \sqrt{E[(X-E[X])^2] E[(Y-E[Y])^2]}\\
\ \ \ |\mbox{Cov}(X,Y)| \le \sqrt{\mbox{Var}(X)\mbox{Var}(Y)}\\
\bullet \text{for correlation coefficient:}\\
\ \ \ |\rho_{XY}| = \frac{|\text{Cov}(X,Y)|}{\sqrt{\text{Var}(X)\text{Var}(Y)}} \le 1\ (\mbox{i.e.},\ -1\le \rho_{XY}\le 1)\\
\bullet \text{$\rho_{XY}=1$ if and only if $Y=aX+b$ for $a>0$}\\
\bullet \text{$\rho_{XY}=-1$ if and only if $Y=aX+b$ for $a<0$}
$

$
\text{a) Cov}(X_1+X_2, X_1+X_3) = \mbox{Cov}(X_1,X_1) + \mbox{Cov}(X_1,X_3) + \mbox{Cov}(X_2,X_3) + \mbox{Cov}(X_2,X_1)
= 5 + 0 + 2 + 2 = 9\\
\text{b) Cov}(X_2-aX_1, X_1) = 0 \implies \mbox{Cov}(X_2,X_1) - a\mbox{Cov}(X_1,X_1) = 2-a\cdot 5 = 0 \implies a = \frac{2}{5}\\
\text{c)}\ \rho_{X_1,X_2} = \frac{\text{Cov}(X_1,X_2)}{\sqrt{\text{Var}(X_1)\text{Var}(X_2)}} = \frac{2}{\sqrt{5\cdot 5}} = \frac{2}{5}\\
\text{d) Var}(X_1+X_2+X_3) = \mbox{Cov}(X_1+X_2+X_3, X_1+X_2+X_3)\\
\ \ \ \ = \mbox{Cov}(X_1,X_1) + \mbox{Cov}(X_2,X_2) + \mbox{Cov}(X_3,X_3)
+ 2\mbox{Cov}(X_1,X_2) + 2\mbox{Cov}(X_2,X_3) + 2\mbox{Cov}(X_1,X_3)\\
= \underbrace{5+5+5+0+0+2+2+2+2}_{\text{sum all entries on matrix}} = 23
